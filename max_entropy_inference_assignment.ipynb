Consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$. The quantum mechanical energy eigenvalues are given by:

Show that the canonical partition function $Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$ forms a convergent geometric series that simplifies to:

2. **Thermodynamic Potentials:**
Using your expression for $Z(T)$, derive explicit formulas for:

3. **Entropy:**
Find the **Entropy** $S(T)$ of the system using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$, and verify that your result satisfies $U = F + TS$.

# Answer: Statistical Mechanics of the Quantum Oscillator

Given energy levels:

$$
E_j=\hbar \omega \left(j+\frac{1}{2}\right), \qquad j=0,1,2,\dots
$$

Let

$$
\beta=\frac{1}{k_B T}.
$$

Then

$$
E_j=\hbar \omega \left(j+\frac{1}{2}\right).
$$

---

## 1. Partition Function

The canonical partition function is

$$
Z(T)=\sum_{j=0}^{\infty} e^{-E_j/(k_B T)}.
$$

Using \(\beta=\frac{1}{k_B T}\),

$$
Z=\sum_{j=0}^{\infty} e^{-\beta E_j}.
$$

Substitute the energy eigenvalues:

$$
Z=\sum_{j=0}^{\infty} e^{-\beta \hbar \omega \left(j+\frac{1}{2}\right)}.
$$

Separate the constant term:

$$
Z=e^{-\frac{\beta \hbar \omega}{2}}
\sum_{j=0}^{\infty} e^{-\beta \hbar \omega j}.
$$

This is a geometric series:

$$
1+r+r^2+r^3+\cdots
$$

where

$$
r=e^{-\beta \hbar \omega}.
$$

Since \(T>0\), we have

$$
0<e^{-\beta \hbar \omega}<1.
$$

Therefore, the geometric series converges.

So,

$$
\sum_{j=0}^{\infty} e^{-\beta \hbar \omega j}
=
\frac{1}{1-e^{-\beta \hbar \omega}}.
$$

Hence,

$$
Z=
\frac{
e^{-\frac{\beta \hbar \omega}{2}}
}{
1-e^{-\beta \hbar \omega}
}.
$$

Since

$$
\beta=\frac{1}{k_B T},
$$

we get

$$
\boxed{
Z(T)=
\frac{
e^{-\frac{\hbar \omega}{2k_B T}}
}{
1-e^{-\frac{\hbar \omega}{k_B T}}
}
}
$$

---

## 2. Thermodynamic Potentials

We have

$$
Z=
\frac{
e^{-\frac{\beta \hbar \omega}{2}}
}{
1-e^{-\beta \hbar \omega}
}.
$$

Taking natural logarithm,

$$
\ln Z
=
-\frac{\beta \hbar \omega}{2}
-
\ln\left(1-e^{-\beta \hbar \omega}\right).
$$

---

### Helmholtz Free Energy

The Helmholtz free energy is

$$
F=-k_B T \ln Z.
$$

Therefore,

$$
F
=
-k_B T
\left[
-\frac{\hbar \omega}{2k_B T}
-
\ln\left(1-e^{-\frac{\hbar \omega}{k_B T}}\right)
\right].
$$

So,

$$
\boxed{
F(T)
=
\frac{\hbar \omega}{2}
+
k_B T
\ln\left(
1-e^{-\frac{\hbar \omega}{k_B T}}
\right)
}
$$

---

### Internal Energy

The internal energy is

$$
U=-\frac{\partial \ln Z}{\partial \beta}.
$$

Now,

$$
\ln Z
=
-\frac{\beta \hbar \omega}{2}
-
\ln\left(1-e^{-\beta \hbar \omega}\right).
$$

Differentiate with respect to \(\beta\):

$$
\frac{\partial \ln Z}{\partial \beta}
=
-\frac{\hbar \omega}{2}
-
\frac{
\hbar \omega e^{-\beta \hbar \omega}
}{
1-e^{-\beta \hbar \omega}
}.
$$

Therefore,

$$
U
=
-\frac{\partial \ln Z}{\partial \beta}.
$$

So,

$$
U
=
\frac{\hbar \omega}{2}
+
\frac{
\hbar \omega e^{-\beta \hbar \omega}
}{
1-e^{-\beta \hbar \omega}
}.
$$

This can be written as

$$
\boxed{
U(T)
=
\frac{\hbar \omega}{2}
+
\frac{\hbar \omega}
{
e^{\frac{\hbar \omega}{k_B T}}-1
}
}
$$

The first term,

$$
\frac{\hbar \omega}{2},
$$

is called the zero-point energy.

---

## 3. Entropy

Entropy is given by

$$
S=-\left(\frac{\partial F}{\partial T}\right).
$$

Let

$$
x=\frac{\hbar \omega}{k_B T}.
$$

Then the free energy becomes

$$
F=
\frac{\hbar \omega}{2}
+
k_B T \ln(1-e^{-x}).
$$

Differentiate with respect to \(T\):

$$
\frac{\partial F}{\partial T}
=
k_B \ln(1-e^{-x})
-
k_B \frac{x}{e^x-1}.
$$

Therefore,

$$
S=
-\frac{\partial F}{\partial T}.
$$

Hence,

$$
\boxed{
S
=
-k_B \ln(1-e^{-x})
+
k_B \frac{x}{e^x-1}
}
$$

where

$$
x=\frac{\hbar \omega}{k_B T}.
$$

Therefore,

$$
\boxed{
S(T)
=
-k_B
\ln\left(
1-e^{-\frac{\hbar \omega}{k_B T}}
\right)
+
k_B
\frac{
\frac{\hbar \omega}{k_B T}
}{
e^{\frac{\hbar \omega}{k_B T}}-1
}
}
$$

---

### Verification of \(U=F+TS\)

We know

$$
F=
\frac{\hbar \omega}{2}
+
k_B T \ln(1-e^{-x})
$$

and

$$
S=
-k_B \ln(1-e^{-x})
+
k_B \frac{x}{e^x-1}.
$$

Then

$$
F+TS
=
\frac{\hbar \omega}{2}
+
k_B T \ln(1-e^{-x})
+
T
\left[
-k_B \ln(1-e^{-x})
+
k_B \frac{x}{e^x-1}
\right].
$$

The logarithm terms cancel:

$$
F+TS
=
\frac{\hbar \omega}{2}
+
k_B T \frac{x}{e^x-1}.
$$

Since

$$
x=\frac{\hbar \omega}{k_B T},
$$

we have

$$
k_B T x=\hbar \omega.
$$

Therefore,

$$
F+TS
=
\frac{\hbar \omega}{2}
+
\frac{\hbar \omega}{e^x-1}.
$$

Thus,

$$
\boxed{
F+TS=U
}
$$

---

## 4. Heat Capacity

The heat capacity at constant volume is

$$
C_V=\left(\frac{\partial U}{\partial T}\right).
$$

We have

$$
U=
\frac{\hbar \omega}{2}
+
\frac{\hbar \omega}{e^x-1},
$$

where

$$
x=\frac{\hbar \omega}{k_B T}.
$$

The zero-point energy term is constant, so its derivative is zero.

Now,

$$
\frac{dx}{dT}
=
-\frac{x}{T}.
$$

Therefore,

$$
C_V
=
\hbar \omega
\frac{d}{dT}
\left(
\frac{1}{e^x-1}
\right).
$$

Using the chain rule,

$$
\frac{d}{dT}
\left(
\frac{1}{e^x-1}
\right)
=
-\frac{e^x}{(e^x-1)^2}
\frac{dx}{dT}.
$$

Since

$$
\frac{dx}{dT}=-\frac{x}{T},
$$

we get

$$
C_V
=
\hbar \omega
\frac{x}{T}
\frac{e^x}{(e^x-1)^2}.
$$

Since

$$
\frac{\hbar \omega}{T}=k_B x,
$$

we obtain

$$
\boxed{
C_V
=
k_B x^2
\frac{e^x}{(e^x-1)^2}
}
$$

where

$$
x=\frac{\hbar \omega}{k_B T}.
$$

Therefore,

$$
\boxed{
C_V
=
k_B
\left(
\frac{\hbar \omega}{k_B T}
\right)^2
\frac{
e^{\frac{\hbar \omega}{k_B T}}
}{
\left(
e^{\frac{\hbar \omega}{k_B T}}-1
\right)^2
}
}
$$

---

## Hyperbolic Form of Heat Capacity

We start with

$$
C_V=
k_B x^2
\frac{e^x}{(e^x-1)^2}.
$$

Now,

$$
e^x-1
=
e^{x/2}
\left(
e^{x/2}-e^{-x/2}
\right).
$$

Therefore,

$$
(e^x-1)^2
=
e^x
\left(
e^{x/2}-e^{-x/2}
\right)^2.
$$

So,

$$
\frac{e^x}{(e^x-1)^2}
=
\frac{1}
{
\left(
e^{x/2}-e^{-x/2}
\right)^2
}.
$$

Using

$$
\sinh y=\frac{e^y-e^{-y}}{2},
$$

put

$$
y=\frac{x}{2}.
$$

Then,

$$
e^{x/2}-e^{-x/2}
=
2\sinh\left(\frac{x}{2}\right).
$$

So,

$$
\left(
e^{x/2}-e^{-x/2}
\right)^2
=
4\sinh^2\left(\frac{x}{2}\right).
$$

Therefore,

$$
\frac{e^x}{(e^x-1)^2}
=
\frac{1}{4\sinh^2(x/2)}.
$$

Substitute into \(C_V\):

$$
C_V=
k_B x^2
\frac{1}{4\sinh^2(x/2)}.
$$

Hence,

$$
C_V=
k_B
\left(\frac{x}{2}\right)^2
\frac{1}{\sinh^2(x/2)}.
$$

Since

$$
x=\frac{\hbar \omega}{k_B T},
$$

we finally get

$$
\boxed{
C_V
=
k_B
\left(
\frac{\hbar \omega}{2k_B T}
\right)^2
\frac{1}
{
\sinh^2
\left(
\frac{\hbar \omega}{2k_B T}
\right)
}
}
$$

---

## 5. High-Temperature Limit

In the high-temperature limit,

$$
k_B T \gg \hbar \omega.
$$

This means

$$
x=\frac{\hbar \omega}{k_B T}\ll 1.
$$

For small \(x\),

$$
\sinh\left(\frac{x}{2}\right)
\approx
\frac{x}{2}.
$$

So,

$$
C_V
=
k_B
\left(\frac{x}{2}\right)^2
\frac{1}{\sinh^2(x/2)}
\approx
k_B
\left(\frac{x}{2}\right)^2
\frac{1}{(x/2)^2}.
$$

Therefore,

$$
\boxed{
C_V \to k_B
}
$$

as

$$
T\to \infty.
$$

This agrees with the classical equipartition theorem.

For a classical one-dimensional harmonic oscillator, there are two quadratic energy terms:

1. Kinetic energy
2. Potential energy

Each contributes

$$
\frac{1}{2}k_B T.
$$

Therefore,

$$
U_{\text{classical}}=k_B T.
$$

Hence,

$$
C_V=
\frac{\partial U}{\partial T}
=
k_B.
$$

So the quantum result becomes the classical result at high temperature.

$$
\boxed{
\text{At high temperature, } C_V \to k_B.
}
$$

# Q. The Maximum Entropy Distribution under Mean and Variance Constraints

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

# Maximum Entropy Distribution under Mean and Variance Constraints

We are given a continuous random variable

$$
X \in (-\infty,\infty)
$$

with constraints

$$
\mathbb{E}[X]=\mu
$$

and

$$
\mathbb{E}\left[(X-\mu)^2\right]=\sigma^2.
$$

We want to show that the probability density function that maximizes entropy is the normal distribution:

$$
f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left[
-\frac{(x-\mu)^2}{2\sigma^2}
\right].
$$

---

## Step 1: Entropy of a Continuous Random Variable

For a continuous random variable with density \(f(x)\), the entropy is

$$
H(f)=-\int_{-\infty}^{\infty} f(x)\ln f(x)\,dx.
$$

We want to maximize this entropy subject to the following constraints:

$$
\int_{-\infty}^{\infty} f(x)\,dx=1
$$

$$
\int_{-\infty}^{\infty} x f(x)\,dx=\mu
$$

$$
\int_{-\infty}^{\infty} (x-\mu)^2 f(x)\,dx=\sigma^2.
$$

---

## Step 2: Use Lagrange Multipliers

We form the functional

$$
\mathcal{L}
=
-\int_{-\infty}^{\infty} f(x)\ln f(x)\,dx
-\lambda_0
\left[
\int_{-\infty}^{\infty} f(x)\,dx-1
\right]
-\lambda_1
\left[
\int_{-\infty}^{\infty} (x-\mu)f(x)\,dx
\right]
-\lambda_2
\left[
\int_{-\infty}^{\infty} (x-\mu)^2f(x)\,dx-\sigma^2
\right].
$$

Here,

$$
\lambda_0,\lambda_1,\lambda_2
$$

are Lagrange multipliers.

---

## Step 3: Differentiate with Respect to \(f(x)\)

To maximize entropy, set the variation with respect to \(f(x)\) equal to zero.

The derivative of

$$
-f(x)\ln f(x)
$$

with respect to \(f(x)\) is

$$
-(\ln f(x)+1).
$$

Therefore,

$$
-(\ln f(x)+1)
-\lambda_0
-\lambda_1(x-\mu)
-\lambda_2(x-\mu)^2
=0.
$$

So,

$$
\ln f(x)
=
-1-\lambda_0-\lambda_1(x-\mu)-\lambda_2(x-\mu)^2.
$$

Taking exponential on both sides,

$$
f(x)
=
e^{-1-\lambda_0}
e^{-\lambda_1(x-\mu)}
e^{-\lambda_2(x-\mu)^2}.
$$

Let

$$
A=e^{-1-\lambda_0}.
$$

Then

$$
f(x)=A e^{-\lambda_1(x-\mu)}e^{-\lambda_2(x-\mu)^2}.
$$

---

## Step 4: Use the Mean Constraint

Since the distribution is centered at \(\mu\), the term involving \((x-\mu)\) must vanish.

Therefore,

$$
\lambda_1=0.
$$

So the density becomes

$$
f(x)=A e^{-\lambda_2(x-\mu)^2}.
$$

This is already the shape of a Gaussian distribution.

---

## Step 5: Find \(A\) and \(\lambda_2\)

We need to satisfy

$$
\int_{-\infty}^{\infty} f(x)\,dx=1.
$$

That means

$$
\int_{-\infty}^{\infty} A e^{-\lambda_2(x-\mu)^2}\,dx=1.
$$

Using the Gaussian integral,

$$
\int_{-\infty}^{\infty} e^{-a(x-\mu)^2}\,dx
=
\sqrt{\frac{\pi}{a}},
$$

we get

$$
A\sqrt{\frac{\pi}{\lambda_2}}=1.
$$

Therefore,

$$
A=\sqrt{\frac{\lambda_2}{\pi}}.
$$

Now use the variance constraint:

$$
\mathbb{E}\left[(X-\mu)^2\right]=\sigma^2.
$$

For the density

$$
f(x)=A e^{-\lambda_2(x-\mu)^2},
$$

the variance is

$$
\sigma^2=\frac{1}{2\lambda_2}.
$$

Therefore,

$$
\lambda_2=\frac{1}{2\sigma^2}.
$$

Substitute this into \(A\):

$$
A=
\sqrt{
\frac{1}{2\sigma^2\pi}
}.
$$

So,

$$
A=
\frac{1}{\sqrt{2\pi\sigma^2}}.
$$

---

## Step 6: Final Maximum Entropy Density

Substitute

$$
A=\frac{1}{\sqrt{2\pi\sigma^2}}
$$

and

$$
\lambda_2=\frac{1}{2\sigma^2}
$$

into

$$
f(x)=A e^{-\lambda_2(x-\mu)^2}.
$$

Therefore,

$$
f(x)
=
\frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left[
-\frac{(x-\mu)^2}{2\sigma^2}
\right].
$$

Hence,

$$
\boxed{
X\sim N(\mu,\sigma^2)
}
$$

or equivalently,

$$
\boxed{
f(x)=
\frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left[
-\frac{(x-\mu)^2}{2\sigma^2}
\right]
}
$$

---

## Conclusion

The probability distribution that maximizes entropy among all continuous distributions on

$$
(-\infty,\infty)
$$

with fixed mean \(\mu\) and fixed variance \(\sigma^2\) is the normal distribution.

Therefore,

$$
\boxed{
\text{Maximum entropy distribution with fixed mean and variance is Gaussian.}
}
$$

# Q. Multiclass Logistic Regression: Automatic Classification of Emails

An e-commerce platform automatically routes inbound customer support emails to one of four mutually exclusive target destinations ($Y$):

* $Y = 0$ : **Finance**
* $Y = 1$ : **Technical**

Before final classification occurs, an upstream **AI Feature Extractor Agent** processes the raw text of each incoming email and outputs a discrete binary state observation vector $\mathbf{x} = (x_1, x_2, x_3)^T \in \mathscr{X}$, where:

* $x_1 = \mathbb{I}(\text{Financial context detected})$ (e.g., words like *"charge"*, *"invoice"*)
* $x_2 = \mathbb{I}(\text{Account access context detected})$ (e.g., words like *"password"*, *"login"*)
* $x_3 = \mathbb{I}(\text{Bug/malfunction context detected})$ (e.g., words like *"error"*, *"frozen"*)

---

Because human language inherently overlaps, the AI Agent's outputs are distributed non-deterministically across three common operational scenarios. The marginal probability of observing each state is fixed by historical logs to:

1. **Scenario A (Pure Financial):** $\mathbf{x}_A = (1, 0, 0)^T \implies P(\mathbf{x}_A) = 0.40$
2. **Scenario B (Pure Acces):** $\mathbf{x}_B = (0, 1, 0)^T \implies P(\mathbf{x}_B) = 0.25$
3. **Scenario C (Pure Technical):** $\mathbf{x}_C = (0, 0, 1)^T \implies P(\mathbf{x}_C) = 0.05$
3. **Scenario D (Access Technical):** $\mathbf{x}_D = (0, 1, 1)^T \implies P(\mathbf{x}_D) = 0.15$
3. **Scenario E (Billing Access):** $\mathbf{x}_E = (1, 1, 0)^T \implies P(\mathbf{x}_E) = 0.05$

---

Historical manual audits provide the following observed joint dataset expectations:

* Constraint 1: $\mathbb{E}[x_1 \cdot \mathbb{I}(Y=1)] = \mathbf{0.05}$
(Very low. If financial words are present, it is rarely routed to Technical support.)

* Constraint 2: $\mathbb{E}[x_2 \cdot \mathbb{I}(Y=1)] = \mathbf{0.15}$
(Moderate. Account access issues split down the middle depending on context.)

* Constraint 3: $\mathbb{E}[x_3 \cdot \mathbb{I}(Y=1)] = \mathbf{0.16}$
(High. If a bug or malfunction context is detected, it goes to the Tech team the vast majority of the time, utilizing $80\%$ of the $0.20$ max capacity).




# Maximum Entropy Routing Distribution for Customer Support Emails

We are given two routing classes:

$$
Y=0 \quad \text{Finance}
$$

$$
Y=1 \quad \text{Technical}
$$

The AI feature vector is

$$
\mathbf{x}=(x_1,x_2,x_3)^T
$$

where

$$
x_1=\mathbb{I}(\text{Financial context detected})
$$

$$
x_2=\mathbb{I}(\text{Account access context detected})
$$

$$
x_3=\mathbb{I}(\text{Bug/malfunction context detected})
$$

---

## Given Scenarios

| Scenario | Feature vector | Probability |
|---|---|---:|
| A | \((1,0,0)^T\) | \(0.40\) |
| B | \((0,1,0)^T\) | \(0.25\) |
| C | \((0,0,1)^T\) | \(0.05\) |
| D | \((0,1,1)^T\) | \(0.15\) |
| E | \((1,1,0)^T\) | \(0.05\) |

The total probability of the listed scenarios is

$$
0.40+0.25+0.05+0.15+0.05=0.90.
$$

So there is an unlisted probability mass of

$$
1-0.90=0.10.
$$

We assume this remaining probability corresponds to a state such as

$$
\mathbf{x}_0=(0,0,0)^T,
$$

which does not affect the given constraints because

$$
x_1=x_2=x_3=0.
$$

---

## Given Constraints

The observed joint expectations are:

$$
\mathbb{E}[x_1\mathbb{I}(Y=1)]=0.05
$$

$$
\mathbb{E}[x_2\mathbb{I}(Y=1)]=0.15
$$

$$
\mathbb{E}[x_3\mathbb{I}(Y=1)]=0.16
$$

Let

$$
q(\mathbf{x})=P(Y=1|\mathbf{x}).
$$

Then

$$
P(Y=0|\mathbf{x})=1-q(\mathbf{x}).
$$

---

# 1. Maximum Entropy Objective

We want the least biased conditional distribution \(P(Y|\mathbf{x})\) satisfying the given constraints.

Therefore, we maximize the conditional entropy:

$$
H(Y|X)
=
-\sum_{\mathbf{x}}P(\mathbf{x})
\sum_{y=0}^{1}
P(y|\mathbf{x})\ln P(y|\mathbf{x}).
$$

Since \(Y\) is binary,

$$
H(Y|X)
=
-\sum_{\mathbf{x}}P(\mathbf{x})
\left[
q(\mathbf{x})\ln q(\mathbf{x})
+
(1-q(\mathbf{x}))\ln(1-q(\mathbf{x}))
\right].
$$

Subject to

$$
\sum_{\mathbf{x}}P(\mathbf{x})x_1q(\mathbf{x})=0.05
$$

$$
\sum_{\mathbf{x}}P(\mathbf{x})x_2q(\mathbf{x})=0.15
$$

$$
\sum_{\mathbf{x}}P(\mathbf{x})x_3q(\mathbf{x})=0.16.
$$

---

# 2. Maximum Entropy Form

Using Lagrange multipliers, the maximum entropy distribution has the logistic form

$$
P(Y=1|\mathbf{x})
=
\frac{
\exp(\lambda_1x_1+\lambda_2x_2+\lambda_3x_3)
}{
1+\exp(\lambda_1x_1+\lambda_2x_2+\lambda_3x_3)
}.
$$

Equivalently,

$$
P(Y=1|\mathbf{x})
=
\frac{1}
{
1+\exp[-(\lambda_1x_1+\lambda_2x_2+\lambda_3x_3)]
}.
$$

Also,

$$
P(Y=0|\mathbf{x})=1-P(Y=1|\mathbf{x}).
$$

So the maximum entropy model is a logistic model.

---

# 3. Set Up the Constraint Equations

Let

$$
q_A=P(Y=1|(1,0,0))
$$

$$
q_B=P(Y=1|(0,1,0))
$$

$$
q_C=P(Y=1|(0,0,1))
$$

$$
q_D=P(Y=1|(0,1,1))
$$

$$
q_E=P(Y=1|(1,1,0)).
$$

Using the given constraints:

---

## Constraint 1

States with \(x_1=1\) are A and E.

# Q. Asset Pricing and Portfolio Optimization.



In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at $\$100$. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.



# Maximum Entropy Distribution for Future Asset Price

Let

$$
S_T=s, \qquad s>0.
$$

We want to find the least-biased continuous probability density function \(p(s)\) for the future asset price \(S_T\), using only the given market constraints.

The constraints are:

$$
\mathbb{E}[S_T]=102
$$

$$
\mathbb{E}[S_T^2]=10600
$$

$$
\mathbb{E}[\ln(S_T)]=4.60
$$

Also,

$$
\int_0^\infty p(s)\,ds=1.
$$

---

## 1. Entropy Objective

The differential entropy of \(S_T\) is

$$
H[p]=-\int_0^\infty p(s)\ln p(s)\,ds.
$$

We want to maximize

$$
H[p]
$$

subject to the constraints:

$$
\int_0^\infty p(s)\,ds=1
$$

$$
\int_0^\infty s p(s)\,ds=102
$$

$$
\int_0^\infty s^2 p(s)\,ds=10600
$$

$$
\int_0^\infty \ln(s)p(s)\,ds=4.60.
$$

---

## 2. Lagrangian Functional

Using Lagrange multipliers, define

$$
\mathcal{L}
=
-\int_0^\infty p(s)\ln p(s)\,ds
-\lambda_0\left(\int_0^\infty p(s)\,ds-1\right)
$$

$$
-\lambda_1\left(\int_0^\infty s p(s)\,ds-102\right)
-\lambda_2\left(\int_0^\infty s^2 p(s)\,ds-10600\right)
$$

$$
-\lambda_3\left(\int_0^\infty \ln(s)p(s)\,ds-4.60\right).
$$

---

## 3. Differentiate with Respect to \(p(s)\)

To maximize entropy, take the functional derivative with respect to \(p(s)\) and set it equal to zero:

$$
\frac{\delta \mathcal{L}}{\delta p(s)}=0.
$$

Now,

$$
\frac{d}{dp}\left[-p\ln p\right]=-(\ln p+1).
$$

Therefore,

$$
-(\ln p(s)+1)-\lambda_0-\lambda_1s-\lambda_2s^2-\lambda_3\ln(s)=0.
$$

So,

$$
\ln p(s)
=
-1-\lambda_0-\lambda_1s-\lambda_2s^2-\lambda_3\ln(s).
$$

Taking exponential on both sides,

$$
p(s)
=
e^{-1-\lambda_0}
e^{-\lambda_1s}
e^{-\lambda_2s^2}
e^{-\lambda_3\ln(s)}.
$$

Since

$$
e^{-\lambda_3\ln(s)}=s^{-\lambda_3},
$$

we get

$$
p(s)=C s^{-\lambda_3}e^{-\lambda_1s-\lambda_2s^2},
\qquad s>0.
$$

---

## 4. Partition Function

To normalize the density, define the partition function

$$
Z(\lambda_1,\lambda_2,\lambda_3)
=
\int_0^\infty
s^{-\lambda_3}
e^{-\lambda_1s-\lambda_2s^2}
\,ds.
$$

Therefore,

$$
C=\frac{1}{Z}.
$$

Hence, the maximum entropy density is

$$
\boxed{
p(s)
=
\frac{1}{Z}
s^{-\lambda_3}
e^{-\lambda_1s-\lambda_2s^2},
\qquad s>0
}
$$

where

$$
Z
=
\int_0^\infty
s^{-\lambda_3}
e^{-\lambda_1s-\lambda_2s^2}
\,ds.
$$

For integrability, we need

$$
\lambda_2>0
$$

and

$$
\lambda_3<1.
$$

---

## 5. Constraint Equations

The Lagrange multipliers are chosen so that the constraints are satisfied.

Using the partition function,

$$
\mathbb{E}[S_T]
=
-\frac{\partial \ln Z}{\partial \lambda_1}
=
102.
$$

Also,

$$
\mathbb{E}[S_T^2]
=
-\frac{\partial \ln Z}{\partial \lambda_2}
=
10600.
$$

And,

$$
\mathbb{E}[\ln(S_T)]
=
-\frac{\partial \ln Z}{\partial \lambda_3}
=
4.60.
$$

Therefore, the multipliers must solve:

$$
\boxed{
-\frac{\partial \ln Z}{\partial \lambda_1}=102
}
$$

$$
\boxed{
-\frac{\partial \ln Z}{\partial \lambda_2}=10600
}
$$

$$
\boxed{
-\frac{\partial \ln Z}{\partial \lambda_3}=4.60
}
$$

---

## 6. Numerical Solution

Solving these equations numerically gives approximately:

$$
\lambda_1=-0.540842
$$

$$
\lambda_2=0.00260217
$$

$$
\lambda_3=0.999989.
$$

The normalization constant is approximately

$$
Z=5.45912\times 10^{11}.
$$

Therefore,

$$
\frac{1}{Z}
=
1.83180\times 10^{-12}.
$$

---

## 7. Final Maximum Entropy Density

Substitute the numerical values into the density:

$$
p(s)
=
\frac{1}{Z}
s^{-\lambda_3}
e^{-\lambda_1s-\lambda_2s^2}.
$$

Hence,

$$
\boxed{
p(s)
=
1.83180\times 10^{-12}
\,
s^{-0.999989}
\exp\left(
0.540842s
-
0.00260217s^2
\right),
\qquad s>0.
}
$$

Therefore,

$$
\boxed{
S_T
\sim
p(s)
=
1.83180\times 10^{-12}
s^{-0.999989}
e^{0.540842s-0.00260217s^2},
\qquad s>0.
}
$$

---

## 8. Verification of Constraints

Using the above density, we get:

$$
\mathbb{E}[S_T]\approx 102
$$

$$
\mathbb{E}[S_T^2]\approx 10600
$$

$$
\mathbb{E}[\ln(S_T)]\approx 4.60.
$$

Thus, the density satisfies all three market constraints.

## 9. Interpretation

The variance is

$$
\text{Var}(S_T)
=
\mathbb{E}[S_T^2]-\left(\mathbb{E}[S_T]\right)^2.
$$

So,

$$
\text{Var}(S_T)
=
10600-102^2.
$$

Since

$$

$$

we get

$$
\text{Var}(S_T)=196.
$$

Therefore,

$$
\boxed{
\text{Standard deviation}=14.
}
$$

The constraint

$$
\mathbb{E}[\ln(S_T)]=4.60
$$

adds downside-skew information.

Because \(\ln(s)\) becomes very negative when \(s\) is close to zero, this constraint forces the distribution to allow stronger downside risk than a simple Gaussian model.

So the final maximum entropy density is not a normal distribution. It is a positive-support exponential-family distribution:

$$
\boxed{
p(s)
\propto
s^{-\lambda_3}
e^{-\lambda_1s-\lambda_2s^2},
\qquad s>0.
}
$$

This is the least-biased distribution consistent with the observed forward price, second moment, and downside logarithmic risk constraint.
