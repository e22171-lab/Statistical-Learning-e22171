# 1. Derivation of the Bias in the MLE Covariance Matrix

The maximum likelihood estimator of the population covariance matrix is

$$
\widehat{\boldsymbol{\Sigma}}_{MLE}
=
\frac{1}{n}
\sum_{i=1}^{n}
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)^T
$$

where

$$
\widehat{\boldsymbol{\mu}}_n
=
\frac{1}{n}
\sum_{i=1}^{n}
\mathbf{X}_i
$$

is the sample mean.

To check whether this estimator is biased, express the deviation from the sample mean using the true population mean $\boldsymbol{\mu}$:

$$
\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n
=
(\mathbf{X}_i-\boldsymbol{\mu})
-
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
$$

Substituting this into the covariance summation gives

$$
\sum_{i=1}^{n}
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)^T
=
\sum_{i=1}^{n}
\left[
(\mathbf{X}_i-\boldsymbol{\mu})
-
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
\right]
\left[
(\mathbf{X}_i-\boldsymbol{\mu})
-
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
\right]^T
$$

Expanding the matrix product,

$$
=
\sum_{i=1}^{n}
(\mathbf{X}_i-\boldsymbol{\mu})
(\mathbf{X}_i-\boldsymbol{\mu})^T
-
\sum_{i=1}^{n}
(\mathbf{X}_i-\boldsymbol{\mu})
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})^T
$$

$$
-
\sum_{i=1}^{n}
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
(\mathbf{X}_i-\boldsymbol{\mu})^T
+
\sum_{i=1}^{n}
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})^T
$$

Since

$$
\sum_{i=1}^{n}
(\mathbf{X}_i-\boldsymbol{\mu})
=
n(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
$$

the expression becomes

$$
=
\sum_{i=1}^{n}
(\mathbf{X}_i-\boldsymbol{\mu})
(\mathbf{X}_i-\boldsymbol{\mu})^T
-
n(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})^T
$$

Now divide by $n$:

$$
\widehat{\boldsymbol{\Sigma}}_{MLE}
=
\frac{1}{n}
\sum_{i=1}^{n}
(\mathbf{X}_i-\boldsymbol{\mu})
(\mathbf{X}_i-\boldsymbol{\mu})^T
-
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})^T
$$

Taking expectation on both sides,

$$
E[\widehat{\boldsymbol{\Sigma}}_{MLE}]
=
\frac{1}{n}
\sum_{i=1}^{n}
E[
(\mathbf{X}_i-\boldsymbol{\mu})
(\mathbf{X}_i-\boldsymbol{\mu})^T
]
-
E[
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})^T
]
$$

By the definition of covariance,

$$
E[
(\mathbf{X}_i-\boldsymbol{\mu})
(\mathbf{X}_i-\boldsymbol{\mu})^T
]
=
\boldsymbol{\Sigma}
$$

Also,

$$
E[
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})^T
]
=
Var(\widehat{\boldsymbol{\mu}}_n)
=
\frac{1}{n}\boldsymbol{\Sigma}
$$

Therefore,

$$
E[\widehat{\boldsymbol{\Sigma}}_{MLE}]
=
\frac{1}{n}
(n\boldsymbol{\Sigma})
-
\frac{1}{n}\boldsymbol{\Sigma}
$$

$$
=
\boldsymbol{\Sigma}
-
\frac{1}{n}\boldsymbol{\Sigma}
$$

$$
=
\left(\frac{n-1}{n}\right)
\boldsymbol{\Sigma}
$$

Hence,

$$
E[\widehat{\boldsymbol{\Sigma}}_{MLE}]
\neq
\boldsymbol{\Sigma}
$$

So, the MLE covariance matrix is biased downward. This happens because the sample mean is used instead of the true population mean, causing the spread of the data to be slightly underestimated.

---

# 2. Bessel's Correction

To remove the bias, the MLE covariance matrix is multiplied by the correction factor

$$
\frac{n}{n-1}
$$

Therefore, the unbiased sample covariance matrix is

$$
\mathbf{S}
=
\frac{n}{n-1}
\widehat{\boldsymbol{\Sigma}}_{MLE}
$$

Substituting the MLE covariance formula,

$$
\mathbf{S}
=
\frac{n}{n-1}
\left[
\frac{1}{n}
\sum_{i=1}^{n}
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)^T
\right]
$$

So,

$$
\mathbf{S}
=
\frac{1}{n-1}
\sum_{i=1}^{n}
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)^T
$$

Now taking expectation,

$$
E[\mathbf{S}]
=
\frac{n}{n-1}
E[\widehat{\boldsymbol{\Sigma}}_{MLE}]
$$

Since

$$
E[\widehat{\boldsymbol{\Sigma}}_{MLE}]
=
\left(\frac{n-1}{n}\right)
\boldsymbol{\Sigma}
$$

we get

$$
E[\mathbf{S}]
=
\frac{n}{n-1}
\left[
\left(\frac{n-1}{n}\right)
\boldsymbol{\Sigma}
\right]
$$

$$
E[\mathbf{S}]
=
\boldsymbol{\Sigma}
$$

Therefore, using $n-1$ instead of $n$ gives an unbiased estimator of the population covariance matrix.

---

# 3. Operational Error Diagnostics

## Type I Error $(\alpha)$

A Type I error is a false alarm.

This happens when the diagnostic system says that the structure is damaged or unstable, even though the structure is actually healthy.

In hypothesis testing, this means rejecting $H_0$ when $H_0$ is true.

## Type II Error $(\beta)$

A Type II error is a missed detection.

This happens when the structure is actually damaged or cracked, but the diagnostic system fails to detect it.

In hypothesis testing, this means failing to reject $H_0$ when $H_1$ is true.

---

# 4. Effect of an Ultra-Conservative Threshold

Consider a very small significance level:

$$
\alpha = 0.0001
$$

When $\alpha$ is reduced to a very small value, the test becomes highly conservative. The rejection region becomes very small, and the test statistic must be extremely large before the system gives an alarm.

## Impact on Statistical Power

Statistical power is

$$
1-\beta
$$

When $\alpha$ becomes very small, the chance of a false alarm decreases. However, the chance of missing real damage increases.

Therefore, $\beta$ increases.

Since power is $1-\beta$, an increase in $\beta$ causes the statistical power to decrease.

So, an ultra-conservative threshold reduces false alarms, but it increases missed detections.

## Impact on Confidence Ellipsoids

The confidence level becomes

$$
1-\alpha
=
1-0.0001
=
0.9999
$$

This means the confidence ellipsoid becomes very large.

A larger ellipsoid includes almost all normal variation in the system. Because of that, even some real structural changes may still fall inside the healthy region.

Therefore, the diagnostic system becomes less sensitive to real damage or degradation.

---

# Final Result

The expectation of the MLE covariance matrix is

$$
E[\widehat{\boldsymbol{\Sigma}}_{MLE}]
=
\left(\frac{n-1}{n}\right)
\boldsymbol{\Sigma}
$$

Therefore, the MLE covariance estimator is biased downward.

The unbiased covariance estimator is

$$
\mathbf{S}
=
\frac{1}{n-1}
\sum_{i=1}^{n}
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)
(\mathbf{X}_i-\widehat{\boldsymbol{\mu}}_n)^T
$$

Also, reducing $\alpha$ to $0.0001$ lowers false alarms, but it increases missed detections and reduces statistical power.

## **Part B: Theoretical Extension (Asymptotic Distributions & Slutsky's Theorem)**

The Lindeberg-Lévy Multivariate Central Limit Theorem establishes that as the sample snapshot size approaches infinity ($n \to \infty$), the standardized difference between the empirical mean and population mean converges in distribution to a Multivariate Gaussian:

$$\sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \xrightarrow{d} \mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma})$$



However, the true population covariance matrix $\boldsymbol{\Sigma}$ is a latent, unobservable parameter.
* State **Slutsky’s Theorem** rigorously.
* Prove how Slutsky's Theorem justifies substituting the empirical sample covariance matrix $\mathbf{S}$ for the latent population matrix $\boldsymbol{\Sigma}$ as $n \to \infty$, thereby operationalizing the practical parametric modeling distribution used in your dashboard:

$$\widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\left(\widehat{\boldsymbol{\mu}}_n, \frac{1}{n}\mathbf{S}\right)$$







### Answer



Let $\{X_n\}$ and $\{Y_n\}$ be sequences of random variables or random vectors.

If

$$
X_n \xrightarrow{d} X
$$

and

$$
Y_n \xrightarrow{p} c
$$

where $c$ is a constant scalar or matrix, then by **Slutsky's Theorem**,

$$
X_n + Y_n \xrightarrow{d} X + c
$$

$$
X_nY_n \xrightarrow{d} cX
$$

and

$$
\frac{X_n}{Y_n} \xrightarrow{d} \frac{X}{c}
$$

provided that

$$
c \neq 0
$$

This theorem is useful because it allows an unknown population quantity to be replaced by a consistent sample estimate in large samples.

---

# Operational Distribution Justification

The Lindeberg-Levy Multivariate Central Limit Theorem states that

$$
\sqrt{n}
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
\xrightarrow{d}
\mathscr{N}
(\mathbf{0},\boldsymbol{\Sigma})
$$

This means that, as the sample size becomes large, the scaled sample mean error follows an approximately multivariate normal distribution.

To standardize the covariance structure, multiply by the inverse square root of the population covariance matrix:

$$
\boldsymbol{\Sigma}^{-1/2}
\left[
\sqrt{n}
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
\right]
\xrightarrow{d}
\mathscr{N}
(\mathbf{0},\mathbf{I})
$$

Here, $\mathbf{I}$ is the identity matrix. Therefore, this transformation converts the random vector into a standard Gaussian form.

---

# Consistency of the Sample Covariance Matrix

The unbiased sample covariance matrix is denoted by $\mathbf{S}$.

By the Weak Law of Large Numbers, $\mathbf{S}$ is a consistent estimator of the true covariance matrix $\boldsymbol{\Sigma}$.

Therefore,

$$
\mathbf{S}
\xrightarrow{p}
\boldsymbol{\Sigma}
$$

Since the inverse square root function is continuous for a positive definite covariance matrix,

$$
\mathbf{S}^{-1/2}
\xrightarrow{p}
\boldsymbol{\Sigma}^{-1/2}
$$

So, for large sample sizes, $\mathbf{S}^{-1/2}$ becomes very close to $\boldsymbol{\Sigma}^{-1/2}$.

---

# Applying Slutsky's Theorem

From the Multivariate Central Limit Theorem,

$$
\sqrt{n}
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
\xrightarrow{d}
\mathscr{N}
(\mathbf{0},\boldsymbol{\Sigma})
$$

Also, from the consistency of the sample covariance matrix,

$$
\mathbf{S}^{-1/2}
\xrightarrow{p}
\boldsymbol{\Sigma}^{-1/2}
$$

Therefore, by Slutsky's Theorem,

$$
\mathbf{S}^{-1/2}
\left[
\sqrt{n}
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
\right]
\xrightarrow{d}
\boldsymbol{\Sigma}^{-1/2}
\mathscr{N}
(\mathbf{0},\boldsymbol{\Sigma})
$$

Since multiplying by $\boldsymbol{\Sigma}^{-1/2}$ standardizes the covariance matrix,

$$
\boldsymbol{\Sigma}^{-1/2}
\mathscr{N}
(\mathbf{0},\boldsymbol{\Sigma})
=
\mathscr{N}
(\mathbf{0},\mathbf{I})
$$

Hence,

$$
\mathbf{S}^{-1/2}
\left[
\sqrt{n}
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
\right]
\xrightarrow{d}
\mathscr{N}
(\mathbf{0},\mathbf{I})
$$

This shows that replacing $\boldsymbol{\Sigma}$ with $\mathbf{S}$ does not change the asymptotic distribution.

---

# Approximate Distribution of the Sample Mean

Since $\mathbf{S}$ is a consistent estimator of $\boldsymbol{\Sigma}$, for large $n$ we can write

$$
\sqrt{n}
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
\approx
\mathscr{N}
(\mathbf{0},\mathbf{S})
$$

Therefore,

$$
\widehat{\boldsymbol{\mu}}_n
\approx
\mathscr{N}
\left(
\boldsymbol{\mu},
\frac{1}{n}\mathbf{S}
\right)
$$

This means the sample mean is approximately normally distributed with mean $\boldsymbol{\mu}$ and covariance matrix $\frac{1}{n}\mathbf{S}$.

---

# Final Result

By the Multivariate Central Limit Theorem,

$$
\sqrt{n}
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
\xrightarrow{d}
\mathscr{N}
(\mathbf{0},\boldsymbol{\Sigma})
$$

Since

$$
\mathbf{S}
\xrightarrow{p}
\boldsymbol{\Sigma}
$$

we have

$$
\mathbf{S}^{-1/2}
\xrightarrow{p}
\boldsymbol{\Sigma}^{-1/2}
$$

Then, by Slutsky's Theorem,

$$
\mathbf{S}^{-1/2}
\left[
\sqrt{n}
(\widehat{\boldsymbol{\mu}}_n-\boldsymbol{\mu})
\right]
\xrightarrow{d}
\mathscr{N}
(\mathbf{0},\mathbf{I})
$$

Therefore, for large sample sizes,

$$
\widehat{\boldsymbol{\mu}}_n
\approx
\mathscr{N}
\left(
\boldsymbol{\mu},
\frac{1}{n}\mathbf{S}
\right)
$$

So, substituting $\mathbf{S}$ for $\boldsymbol{\Sigma}$ is asymptotically valid and does not disturb the stability of the diagnostic boundary.

````python
import numpy as np
import pandas as pd
from typing import Optional, Sequence, Dict, Any
from sklearn.decomposition import FactorAnalysis
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# =====================================================================
# COMBINED SUBSPACE ENGINES (PCA & FA PIPELINES)
# =====================================================================

class DualSubspaceDiagnosticsEngine:
    def __init__(self, df: pd.DataFrame):
        self.df = df

    def compute_empirical_pca(self, columns: Optional[Sequence[str]] = None, show_plot: bool = True) -> Dict[str, Any]:
        """
        Operationalizes the geometric de-correlation framework via PCA.
        Decomposes the unbiased empirical covariance matrix S into its orthogonal
        basis P_hat and renders a clean 2x3 optimization and feature contribution dashboard.
        """
        if self.df is None:
            raise ValueError("Error: No data loaded.")

        if columns is None:
            target_cols = self.df.select_dtypes(include=[np.number]).columns.tolist()
            if 'count' in target_cols: target_cols.remove('count')
            target_cols = list(target_cols)
        else:
            target_cols = list(columns)

        X = self.df[target_cols].copy().dropna().values
        n, m = X.shape

        if n <= m:
            raise ValueError("Sample size n must be larger than feature dimensions m to isolate principal components.")

        # 1. Centering
        mu_hat = np.mean(X, axis=0)
        X_centered = X - mu_hat

        # 2. Compute Unbiased Sample Covariance Matrix S
        S_matrix = np.cov(X, rowvar=False, ddof=1)

        # 3. Spectral Decomposition
        eigenvalues, eigenvectors = np.linalg.eigh(S_matrix)
        idx = np.argsort(eigenvalues)[::-1]
        lambda_hat = eigenvalues[idx]
        P_hat = eigenvectors[:, idx]
        lambda_hat = np.clip(lambda_hat, a_min=1e-15, a_max=None)

        # 4. Variance Information Calculations
        total_variance = np.sum(lambda_hat)
        if total_variance == 0:
            raise ValueError("Total variance is zero.")

        explained_variance_ratio = lambda_hat / total_variance
        cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
        unexplained_variance_ratio = 1.0 - cumulative_variance_ratio

        # 5. Map Realizations to Principal Component Scores
        Z_scores = np.dot(X_centered, P_hat)
        S_Z = np.cov(Z_scores, rowvar=False, ddof=1)

        # 6. Evaluate Statistical Distance Metrics (T^2 and Q)
        k_range = np.arange(1, m)
        mean_T2_vs_k = []
        mean_Q_vs_k = []
        T2_matrix = np.zeros((n, len(k_range)))
        Q_matrix = np.zeros((n, len(k_range)))

        for idx_k, k in enumerate(k_range):
            Z_k = Z_scores[:, :k]
            lambda_k = lambda_hat[:k]
            T2_samples = np.sum((Z_k ** 2) / lambda_k, axis=1)
            T2_matrix[:, idx_k] = T2_samples
            mean_T2_vs_k.append(np.mean(T2_samples))

            Z_residual = Z_scores[:, k:]
            Q_samples = np.sum(Z_residual ** 2, axis=1)
            Q_matrix[:, idx_k] = Q_samples
            mean_Q_vs_k.append(np.mean(Q_samples))

        print(f"\n--- Operationalizing Geometric De-correlation Layer (PCA) ---")
        print(f"Decomposing structural space of {m} features using {n} samples.")
        print(f"Total System Variance (Trace[S]): {total_variance:.4f}")

        # 7. Enhanced 2x3 Diagnostic Dashboard
        if show_plot:
            pc_labels = [f"PC {i+1}" for i in range(m)]
            k_labels = [f"k={k}" for k in k_range]
            sensor_labels = target_cols

            fig = make_subplots(
                rows=2, cols=3,
                horizontal_spacing=0.18,
                vertical_spacing=0.28,
                subplot_titles=(
                    "Feature Loading Matrix |P_hat|",
                    "Component Values (Eigenvalues λ)",
                    "Information Profile (Explained Var.)",
                    "Residual Space (Unexplained Var.)",
                    "Mean Hotelling's T² vs Subspace Size k",
                    "Mean Q Statistic (SPE) vs Subspace Size k"
                )
            )

            # --- ROW 1, COL 1: PCA Loading Heatmap (Left-anchored Colorbar) ---
            fig.add_trace(
                go.Heatmap(
                    z=np.abs(P_hat),
                    x=pc_labels,
                    y=sensor_labels,
                    colorscale='YlOrRd',
                    colorbar=dict(
                        title="Loading Weight",
                        x=-0.12,          # Perfectly isolates colorbar to the left side
                        len=0.38,         # Fits height bounds of row 1
                        y=0.78,           # Centers vertically inside row 1
                        yanchor="middle",
                        xanchor="right",
                        titleside="top"
                    ),
                    showscale=True,
                    showlegend=False
                ),
                row=1, col=1
            )
            fig.update_xaxes(title_text="Principal Axes", row=1, col=1)

            # --- ROW 1, COL 2: Absolute Eigenvalues ---
            fig.add_trace(go.Bar(x=pc_labels, y=lambda_hat, name="Eigenvalue (λ_j)", marker=dict(color='#1f77b4', line=dict(color='black', width=0.5)), legendgroup="eigen"), row=1, col=2)
            fig.update_yaxes(title_text="Variance Magnitude", row=1, col=2)
            fig.update_xaxes(title_text="Principal Axes", row=1, col=2)

            # --- ROW 1, COL 3: Variance Profiles ---
            fig.add_trace(go.Bar(x=pc_labels, y=explained_variance_ratio * 100, name="Marginal Explained", marker=dict(color='#ff7f0e', opacity=0.75), legendgroup="expl"), row=1, col=3)
            fig.add_trace(go.Scatter(x=pc_labels, y=cumulative_variance_ratio * 100, mode='lines+markers', name='Cumulative Captured', line=dict(color='#d62728', width=2.5, dash='dash'), legendgroup="expl"), row=1, col=3)
            fig.update_yaxes(title_text="Captured Structure (%)", range=[-2, 105], row=1, col=3)
            fig.update_xaxes(title_text="Principal Axes", row=1, col=3)

            # --- ROW 2, COL 1: Remaining Noise ---
            fig.add_trace(go.Bar(x=pc_labels, y=unexplained_variance_ratio * 100, name="Remaining Noise", marker=dict(color='#2ca02c', line=dict(color='black', width=0.5)), legendgroup="noise"), row=2, col=1)
            fig.update_yaxes(title_text="Excluded Info (%)", range=[-2, 105], row=2, col=1)
            fig.update_xaxes(title_text="Principal Axes", row=2, col=1)

            # --- ROW 2, COL 2: Hotelling's T^2 ---
            fig.add_trace(go.Scatter(x=k_labels, y=mean_T2_vs_k, mode='lines+markers', name='Mean T²', line=dict(color='#9467bd', width=2.5), marker=dict(size=6, symbol='diamond'), legendgroup="t2"), row=2, col=2)
            fig.update_yaxes(title_text="Average T² Metric", row=2, col=2)
            fig.update_xaxes(title_text="Truncation Cutoff (k)", row=2, col=2)

            # --- ROW 2, COL 3: SPE / Q Statistic ---
            fig.add_trace(go.Scatter(x=k_labels, y=mean_Q_vs_k, mode='lines+markers', name='Mean Q (SPE)', line=dict(color='#e377c2', width=2.5), marker=dict(size=6, symbol='square'), legendgroup="q_stat"), row=2, col=3)
            fig.update_yaxes(title_text="Average Residual Energy", row=2, col=3)
            fig.update_xaxes(title_text="Truncation Cutoff (k)", row=2, col=3)

            # Global Unified Layout Configurations
            fig.update_layout(
                title=dict(text="Principal Component Analysis (PCA) Optimization & Feature Loading Dashboard", x=0.5, y=0.97, xanchor="center", yanchor="top"),
                template="plotly_white",
                showlegend=True,
                legend=dict(
                    orientation="h",
                    yanchor="bottom",
                    y=1.02,
                    xanchor="center",
                    x=0.5
                ),
                margin=dict(t=150, b=60, l=140, r=80),
                height=750,
                width=1250
            )
            fig.show()

        return {
            "mean_vector": mu_hat, "covariance_matrix_S": S_matrix, "eigenvalues_lambda": lambda_hat,
            "eigenvectors_P": P_hat, "explained_variance_ratio": explained_variance_ratio,
            "cumulative_variance_ratio": cumulative_variance_ratio, "unexplained_variance_ratio": unexplained_variance_ratio,
            "transformed_scores_Z": Z_scores, "score_covariance_diagonal": np.diag(S_Z),
            "features": target_cols, "k_values": k_range, "T2_matrix_vs_k": T2_matrix, "Q_matrix_vs_k": Q_matrix,
            "mean_T2_profile": np.array(mean_T2_vs_k), "mean_Q_profile": np.array(mean_Q_vs_k)
        }

    def compute_empirical_fa(self, k: int, columns: Optional[Sequence[str]] = None, show_plot: bool = True) -> Dict[str, Any]:
        """
        Operationalizes the Factor Analysis latent subspace framework with a clean 2x2 grid.
        Anchors the heatmap colorbar perfectly to the right edge of the first subplot.
        """
        if self.df is None:
            raise ValueError("Error: No data loaded.")

        if columns is None:
            target_cols = self.df.select_dtypes(include=[np.number]).columns.tolist()
            if 'count' in target_cols: target_cols.remove('count')
            target_cols = list(target_cols)
        else:
            target_cols = list(columns)

        X = self.df[target_cols].copy().dropna().values
        n, m = X.shape

        # 1. Standardize realizations to construct Z space
        mu_hat = np.mean(X, axis=0)
        std_hat = np.std(X, axis=0, ddof=1)
        std_hat[std_hat == 0] = 1e-15
        Z = (X - mu_hat) / std_hat
        R_matrix = np.corrcoef(X, rowvar=False)

        # 2. Fit Factor Analysis Model (Varimax Rotation)
        fa = FactorAnalysis(n_components=k, rotation='varimax', random_state=42)
        fa.fit(Z)
        lambda_matrix = fa.components_.T
        psi_diagonal = fa.noise_variance_

        # 3. Calculate Communality (h²) and Uniqueness (φ²) vectors
        communality = np.sum(lambda_matrix**2, axis=1)
        uniqueness = psi_diagonal

        # 4. Map realizations using Thomson's Regression Method
        F_scores = fa.transform(Z)

        print(f"\n--- Operationalizing Latent Subspace Layer (Factor Analysis) ---")
        print(f"Decomposing structural space of {m} sensors into {k} hidden physical factors.")
        print(f"Average System Communality (Shared Subspace Energy): {np.mean(communality)*100:.2f}%")
        print(f"Average System Uniqueness (Localized Noise Floor): {np.mean(uniqueness)*100:.2f}%")

        # 5. Clean 2x2 Dashboard
        if show_plot:
            sensor_labels = target_cols
            factor_labels = [f"Factor {j+1}" for j in range(k)]

            fig = make_subplots(
                rows=2, cols=2, horizontal_spacing=0.24, vertical_spacing=0.28,
                subplot_titles=(
                    "Structural Loadings Matrix |λ_(j,r)|",
                    "Variance Partitioning (Communality vs Uniqueness)",
                    "Sensor Uniqueness Noise Floor (φ²)",
                    "Latent Factor Scores Empirical Variance"
                )
            )

            # --- Subplot 1: Factor Loadings Heatmap (Colorbar repositioned to x=0.42) ---
            fig.add_trace(
                go.Heatmap(
                    z=np.abs(lambda_matrix),
                    x=factor_labels,
                    y=sensor_labels,
                    colorscale='YlOrRd',
                    colorbar=dict(
                        title="Sensitivity",
                        x=-0.1,          # Shifts colorbar out of the far right to center-left
                        len=0.38,         # Scales down length to match row 1 height bounds
                        y=0.78,          # Centers vertically inside row 1
                        yanchor="middle",
                        xanchor="right"
                    ),
                    showscale=True,
                    showlegend=False
                ),
                row=1, col=1
            )
            fig.update_xaxes(title_text="Latent Structures", row=1, col=1)

            # --- Subplot 2: Variance Partitioning ---
            fig.add_trace(go.Bar(y=sensor_labels, x=communality * 100, name="Communality (h² - Shared Structure)", orientation='h', marker=dict(color='#1f77b4'), legendgroup="communality"), row=1, col=2)
            fig.add_trace(go.Bar(y=sensor_labels, x=uniqueness * 100, name="Uniqueness (φ² - Channel Noise)", orientation='h', marker=dict(color='#ff7f0e'), legendgroup="uniqueness_bar"), row=1, col=2)
            fig.update_layout(barmode='stack')
            fig.update_xaxes(title_text="Variance Allocation (%)", range=[0, 100], row=1, col=2)

            # --- Subplot 3: Uniqueness Line Profile ---
            fig.add_trace(go.Scatter(x=sensor_labels, y=uniqueness, mode='lines+markers', name='Uniqueness Profile (φ²)', line=dict(color='#d62728', width=2, dash='dot'), marker=dict(size=8, symbol='x'), legendgroup="uniqueness_line"), row=2, col=1)
            fig.update_yaxes(range=[-0.05, 1.05], row=2, col=1)
            fig.update_xaxes(title_text="Monitored Channels", tickangle=25, row=2, col=1)

            # --- Subplot 4: Latent Factor Scores Empirical Variance ---
            factor_variances = np.var(F_scores, axis=0, ddof=1)
            fig.add_trace(go.Bar(x=factor_labels, y=factor_variances, name="Factor Empirical Variance", marker=dict(color='#2ca02c', line=dict(color='black', width=0.5)), legendgroup="variance"), row=2, col=2)
            fig.update_yaxes(title_text="Variance Level", row=2, col=2)
            fig.update_xaxes(title_text="Latent Vectors", row=2, col=2)

            # Global Layout Updates
            fig.update_layout(
                title=dict(text="Factor Analysis (FA) Latent Subspace Diagnostics Dashboard", x=0.5, y=0.97, xanchor="center", yanchor="top"),
                template="plotly_white",
                showlegend=True,
                legend=dict(
                    orientation="h",
                    yanchor="bottom",
                    y=1.02,
                    xanchor="center",
                    x=0.5,
                    traceorder="normal",
                    itemwidth=40,
                    itemsizing="constant"
                ),
                margin=dict(t=100, b=60, l=80, r=80),
                height=750,
                width=1100
            )
            fig.show()

        return {
            "mean_vector_mu": mu_hat, "std_vector_D": std_hat, "correlation_matrix_R": R_matrix,
            "factor_loadings_lambda": lambda_matrix, "uniqueness_psi": uniqueness,
            "communality_h2": communality, "latent_factor_scores_F": F_scores, "sensors": target_cols
        }

# =====================================================================
# DATA GENERATOR & PIPELINE EXECUTION
# =====================================================================
if __name__ == "__main__":
    # Set execution seed for perfect reproducible properties
    np.random.seed(42)
    n_samples = 2500

    # 1. Generate underlying, independent true physical driving forces (Latent Factors)
    f1 = np.random.normal(0, 1, n_samples)  # Primary structural mode/load
    f2 = np.random.normal(0, 1, n_samples)  # Secondary operational mode

    # 2. Construct the Observable Multi-Sensor Array with targeted noise profiles
    # Sensors 1 and 2 couple strongly to the primary process (f1)
    s1 = 0.85 * f1 + 0.10 * f2 + np.random.normal(0, 0.3, n_samples)
    s2 = 0.80 * f1 + 0.15 * f2 + np.random.normal(0, 0.35, n_samples)

    # Sensor 3 couples strongly to the secondary process (f2)
    s3 = 0.12 * f1 + 0.90 * f2 + np.random.normal(0, 0.25, n_samples)

    # Sensor 4 suffers an extreme electrical/instrumentation fault (massive localized white noise)
    s4 = 0.02 * f1 + 0.05 * f2 + np.random.normal(0, 1.40, n_samples)

    # 3. Compile into the final evaluation DataFrame
    df_asset = pd.DataFrame(
        data=np.vstack([s1, s2, s3, s4]).T,
        columns=['Sensor_1', 'Sensor_2', 'Sensor_3', 'Sensor_4']
    )
    # Instantiate Engine
    engine = DualSubspaceDiagnosticsEngine(df_asset)

    # Run PCA Engine (1x5 Dashboard)
    pca_metrics = engine.compute_empirical_pca(show_plot=True)

    # Run FA Engine (2x2 Dashboard, Retaining k=2 Common Physical Factors)
    fa_metrics = engine.compute_empirical_fa(k=2, show_plot=True)
````

### Solution for Question 1: The Total Variance Illusion

* **Part 1 (Physical Meaning of Sensor 4):** A Uniqueness value $(\varphi^2)$ close to $100\%$ means that `Sensor_4` has almost no shared covariance with the other sensors in the system. It is not connected to the main physical factors $(f_1, f_2)$ that explain the behavior of the plant asset. Therefore, the signal from `Sensor_4` is mostly made up of isolated white noise, measurement error, or a serious electrical fault.

* **Part 2 (PCA Loading Contamination):** From the **Feature Loading Matrix** in the PCA dashboard, a clear abnormal pattern can be observed. `PC 1` is almost completely controlled by one variable, which is `Sensor_4`. This is shown by the strong red block with a loading value close to $1.0$. Since `Sensor_4` has a very large noise variance scale $(\sigma^2 \approx 2.0)$, the PCA model captures this large variance and assigns its first principal component mainly to that noisy sensor. As a result, `Sensor_1`, `Sensor_2`, and `Sensor_3` have nearly $0$ loading on `PC 1`.

* **Part 3 (Engineering Risk):** This creates a serious monitoring risk. If standard PCA is used for structural or plant monitoring, one damaged, blown, or short-circuited sensor can dominate the most important monitoring direction, `PC 1`. This increases the first eigenvalue to around $\lambda_1 \approx 2.0$ and hides the true common structural behavior of the asset. Because of this, the main PCA control chart may become ineffective for detecting real system-level faults or structural failures.

### Solution for Question 2: Decoupling Structural Loading via Rotation

* **Part 1 (Mathematical Mixing vs. Clear Structure):** Because PCA is restricted to orthogonal variance maximization, the remaining signals on `PC 2` and `PC 4` become mixed. `PC 2` and `PC 4` show high loading values on both `Sensor_1` and `Sensor_2`, and they also slightly spread into `Sensor_3`. This type of cross-loading creates a forced mathematical mixture of sensor signals. Factor Analysis reduces this problem by applying a **Varimax rotation**. By increasing the separation of loadings within each column, it produces a clearer structure: `Sensor_1` and `Sensor_2` group strongly under `Factor 1`, while `Sensor_3` loads mainly onto `Factor 2`.

* **Part 2 (Operational Interpretability):** The PCA loading matrix gives a more complicated interpretation. If an alarm is triggered on `PC 2`, the engineer must analyze a mixed contribution from `Sensor_1`, `Sensor_2`, and `Sensor_3`. In the rotated Factor Analysis model, the structure is easier to understand. `Factor 1` represents the common physical process shared by `Sensor_1` and `Sensor_2`, while `Factor 2` represents the separate physical process captured by `Sensor_3`. This makes fault diagnosis and root-cause identification much faster and clearer.

### Solution for Question 3: Determining Subspace Truncation ($k$)

* **Part 1 (Curve Observation):** When moving from $k=1$ to $k=2$, the Mean $Q$ Statistic (SPE) drops sharply from about $2.4$ to $0.9$. After that, from $k=2$ to $k=3$, the curve starts to level off and forms a clear "elbow". The decrease becomes much slower and continues only gradually toward $0.1$.

* **Part 2 (True Intrinsic Dimension):** Comparing the $Q$ curve with the loading matrix makes the choice more clear. `PC 1` mainly captures the isolated noise coming from `Sensor_4`. `PC 2` explains the real shared variation among the working sensors, mainly `Sensor_1`, `Sensor_2`, and `Sensor_3`. The large fall in the $Q$ statistic at $k=2$ shows that these two main types of behavior have been captured. Therefore, $k=2$ can be considered the true hidden physical dimension of the asset.

* **Part 3 (Risk of Over-fitting with $k=3$):** In the loading matrix, `PC 3` mainly represents the remaining residual variation of `Sensor_3`. If the model keeps $k=3$, this residual behavior is included inside the normal model space. This can overfit the system and make the Hotelling's $T^2$ control chart too sensitive. As a result, small normal baseline changes may create unnecessary false alarms.

### Solution for Question 4: Operational Trade-offs in System Health Monitoring

* **The PCA Strategy:** Monitoring the main principal component space using $T^2$ together with the residual space using the $Q$ statistic.

* *Vulnerability:* This method is very sensitive to sensor degradation. As seen in the PCA loading matrix, if one sensor channel such as `Sensor_4` fails with a high variance, it can change the direction of the main principal axes. This affects the calculations of the other healthy sensors as well, because the PCA model may start focusing too much on the faulty sensor.


* **The FA Strategy:** Monitoring the rotated latent factor scores directly.

* *Robustness:* This method is more robust. Factor Analysis separates local sensor noise using the diagonal uniqueness matrix $(\Psi)$. Therefore, a failure in one sensor channel can be captured mainly by that sensor’s uniqueness value, such as $\psi_4^2 \rightarrow 1$. Because of this, the main latent factors, `Factor 1` and `Factor 2`, remain clean and stable. The rest of the health monitoring system can continue working normally without being strongly affected by the faulty sensor.
