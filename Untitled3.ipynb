# Gaussian Process Regression

```python
import os
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error


def load_data(filename):
    paths = [filename, f'/content/{filename}', f'/mnt/data/{filename}']
    for path in paths:
        if os.path.exists(path):
            return pd.read_csv(path)

    from google.colab import files
    files.upload()
    return pd.read_csv(filename)
```

Consider the following [data set](https://www.kaggle.com/datasets/elikplim/eergy-efficiency-dataset) that has been created in an energy analysis using 12 different building shapes simulated in Ecotect. The buildings differ with respect to the glazing area, the glazing area distribution, and the orientation, amongst other parameters. The dataset contains eight attributes (or features, denoted by X1 to X8) and two responses (denoted by Y1 and Y2). Explore the possibility of modeling the 'heating load' and the 'cooling load' as a single parameter Gaussian process. Discuss your conclusions.

##Q1 Answer

### Load and inspect the dataset

```python
energy_df = load_data('ENB2012_data.csv')

print('Dataset shape:', energy_df.shape)
print('Missing values:', energy_df.isnull().sum().sum())
energy_df.head()
```

### Check the relationship between heating and cooling loads

If `Y1` and `Y2` are strongly related, they can be represented using one common parameter.

```python
correlation = energy_df['Y1'].corr(energy_df['Y2'])
print('Correlation between Y1 and Y2:', round(correlation, 4))

plt.figure(figsize=(6, 4))
plt.scatter(energy_df['Y1'], energy_df['Y2'], alpha=0.6)
plt.xlabel('Heating Load (Y1)')
plt.ylabel('Cooling Load (Y2)')
plt.title('Heating Load vs Cooling Load')
plt.grid(alpha=0.3)
plt.show()
```

The correlation is very high. Therefore, Principal Component Analysis (PCA) is used to combine `Y1` and `Y2` into one latent parameter.

### Split and standardize the data

```python
features = [f'X{i}' for i in range(1, 9)]

X = energy_df[features]
y = energy_df[['Y1', 'Y2']]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

x_scaler = StandardScaler()
y_scaler = StandardScaler()

X_train_scaled = x_scaler.fit_transform(X_train)
X_test_scaled = x_scaler.transform(X_test)

y_train_scaled = y_scaler.fit_transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

### Create one parameter using PCA

```python
pca = PCA(n_components=1)

z_train = pca.fit_transform(y_train_scaled).ravel()
z_test = pca.transform(y_test_scaled).ravel()

print('Variance retained by the single parameter:',
      round(pca.explained_variance_ratio_[0] * 100, 2), '%')
```

### Train the Gaussian Process model

```python
gpr_model = GaussianProcessRegressor(
    kernel=RBF(length_scale=1.0),
    alpha=0.05,
    normalize_y=True,
    optimizer=None,
    random_state=42
)

gpr_model.fit(X_train_scaled, z_train)
z_pred = gpr_model.predict(X_test_scaled)

# Convert the predicted single parameter back to Y1 and Y2
y_pred_scaled = pca.inverse_transform(z_pred.reshape(-1, 1))
y_pred = y_scaler.inverse_transform(y_pred_scaled)
```

### Evaluate the predictions

```python
gpr_results = pd.DataFrame({
    'Output': ['Heating Load', 'Cooling Load'],
    'R2': [
        r2_score(y_test['Y1'], y_pred[:, 0]),
        r2_score(y_test['Y2'], y_pred[:, 1])
    ],
    'RMSE': [
        mean_squared_error(y_test['Y1'], y_pred[:, 0]) ** 0.5,
        mean_squared_error(y_test['Y2'], y_pred[:, 1]) ** 0.5
    ],
    'MAE': [
        mean_absolute_error(y_test['Y1'], y_pred[:, 0]),
        mean_absolute_error(y_test['Y2'], y_pred[:, 1])
    ]
})

gpr_results.round(4)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))

axes[0].scatter(y_test['Y1'], y_pred[:, 0], alpha=0.6)
axes[0].plot([y_test['Y1'].min(), y_test['Y1'].max()],
             [y_test['Y1'].min(), y_test['Y1'].max()], '--')
axes[0].set_xlabel('Actual Heating Load')
axes[0].set_ylabel('Predicted Heating Load')
axes[0].set_title('Heating Load')
axes[0].grid(alpha=0.3)

axes[1].scatter(y_test['Y2'], y_pred[:, 1], alpha=0.6)
axes[1].plot([y_test['Y2'].min(), y_test['Y2'].max()],
             [y_test['Y2'].min(), y_test['Y2'].max()], '--')
axes[1].set_xlabel('Actual Cooling Load')
axes[1].set_ylabel('Predicted Cooling Load')
axes[1].set_title('Cooling Load')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

# Modeling Heating Load and Cooling Load as a Single-Parameter Gaussian Process

The Energy Efficiency dataset contains eight input features $(X_1)$–$(X_8)$ that describe building characteristics and two response variables: Heating Load $(Y_1)$ and Cooling Load $(Y_2)$. The main purpose is to examine whether these two outputs can be represented using one combined parameter and modeled with a Gaussian Process (GP).

A basic analysis of the dataset shows that heating load and cooling load are highly correlated. Buildings with higher heating loads usually also show higher cooling loads. This indicates that both responses are affected by similar building features such as relative compactness, surface area, wall area, roof area, glazing area, glazing area distribution, and orientation.

The relationship between the two response variables can be measured using the Pearson correlation coefficient:

$$
\rho_{Y_1,Y_2}
=
\frac{\text{Cov}(Y_1,Y_2)}
{\sigma_{Y_1}\sigma_{Y_2}}
$$

For this dataset, the correlation coefficient is approximately $0.98$, which shows a very strong positive relationship between heating load and cooling load.

Because of this strong relationship, it is possible to define a single latent response variable:

$$
Z = \frac{Y_1 + Y_2}{2}
$$

Then this combined variable can be modeled using a Gaussian Process:

$$
Z = f(\mathbf{x}) + \varepsilon
$$

where

$$
\mathbf{x} = (X_1, X_2, \ldots, X_8)
$$

represents the building input features, and

$$
\varepsilon \sim \mathcal{N}(0,\sigma^2)
$$

represents Gaussian noise.

Using a single-output Gaussian Process has some advantages. It reduces the dimensionality of the output space, makes the model simpler, and uses the strong correlation between the two energy load responses. Since heating load and cooling load follow similar patterns, one latent variable may capture most of the important information from both outputs.

However, heating load and cooling load are not exactly the same physical quantity. They represent different energy requirements and may react differently to changes in building design. For example, increasing glazing area may have a stronger effect on cooling demand than on heating demand. Therefore, combining both responses into one parameter may remove some useful information specific to each output.

A better option is to use a multi-output Gaussian Process model, because it can model heating load and cooling load together while still keeping their individual meanings. This type of model can use the correlation between the two responses without forcing them into one single value.

## Conclusion

The strong correlation between heating load and cooling load suggests that they can be approximately represented by a single latent parameter and modeled using a single-output Gaussian Process. This method is useful for reducing model complexity and capturing the main trend in the data. However, since heating load and cooling load are still different responses and may behave differently under some building features, a multi-output Gaussian Process or separate Gaussian Process models are usually more accurate. Therefore, a single-parameter Gaussian Process is possible as a simple approximation, but it is not the best modeling approach for this dataset.

# Linear Regression

Consider the following [data set](https://www.kaggle.com/datasets/programmer3/green-building-multi-source-environment-dataset). This dataset has 2400 samples provides a comprehensive collection of multi-source building environment data designed to support research in green building design, energy efficiency optimization, and indoor comfort prediction using advanced machine learning and deep learning techniques. Explore the possibility of predicting the 'predicted_energy_demand'  using a linear relationship of a suitable set of other data parameters. Justify your choice of parameters and discuss the results.

## Q2 Answer

### Load and inspect the dataset

```python
green_df = load_data('green_building_dataset.csv')

print('Dataset shape:', green_df.shape)
print('Missing values:', green_df.isnull().sum().sum())
green_df.head()
```

### Check correlations and select parameters

```python
target = 'predicted_energy_demand'

correlations = green_df.corr(numeric_only=True)[target].drop(target)
correlations = correlations.sort_values(key=abs, ascending=False)

correlations.head(10).round(4)
```

The selected parameters are:

- `ventilation_rate`
- `electricity_consumption`
- `cooling_energy`
- `heating_energy`
- `equipment_load`
- `occupancy`

The first four variables have the strongest relationships with energy demand. Equipment load and occupancy are also included because they directly contribute to building energy use. `predicted_comfort_index` is not used because it is another predicted output.

### Split the data and train the linear regression model

```python
selected_features = [
    'ventilation_rate',
    'electricity_consumption',
    'cooling_energy',
    'heating_energy',
    'equipment_load',
    'occupancy'
]

X = green_df[selected_features]
y = green_df[target]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred = lr_model.predict(X_test)
```

### Evaluate the model

```python
r2 = r2_score(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred) ** 0.5
mae = mean_absolute_error(y_test, y_pred)

print('R2:', round(r2, 4))
print('RMSE:', round(rmse, 4))
print('MAE:', round(mae, 4))
```

### Display the fitted equation

```python
coefficient_table = pd.DataFrame({
    'Parameter': selected_features,
    'Coefficient': lr_model.coef_
})

print('Intercept:', round(lr_model.intercept_, 5))
coefficient_table.round(5)
```

```python
equation = f'{lr_model.intercept_:.5f}'
for feature, coefficient in zip(selected_features, lr_model.coef_):
    equation += f' + ({coefficient:.5f} × {feature})'

print('Predicted Energy Demand =')
print(equation)
```

### Compare actual and predicted values

```python
plt.figure(figsize=(6, 4))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()], '--')
plt.xlabel('Actual Energy Demand')
plt.ylabel('Predicted Energy Demand')
plt.title('Actual vs Predicted Energy Demand')
plt.grid(alpha=0.3)
plt.show()
```

### Conclusion

The linear regression model obtained an \(R^2\) value of approximately **0.961**, which means that it explains about **96.1%** of the variation in `predicted_energy_demand`. The RMSE is approximately **1.901** and the MAE is approximately **1.495**.

Therefore, a linear relationship using the six selected parameters is suitable for predicting energy demand in this dataset. Ventilation rate, electricity consumption, cooling energy, and heating energy are the main predictors, while equipment load and occupancy provide additional information.
