#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{ \omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.

4. Compute $E[X\mid Y]$ explicitly and show that

$$
E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}
$$

5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$,

$$
\int_A E[X\mid Y]\,dP=\int_A X\,dP.
$$

6. Find the Conditional PMF $p_{X|Y}(x \mid y)$.

7. Verify the identity:
$$\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$$.

8. Calculate the **Mean Square Error (MSE)**

9. How would the **Mean Square Error (MSE)** change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$) as a constant estimate for every transaction? Calculate this error and explain its relationship to the **Law of Total Variance**.

10. Using the joint distribution of $(X, Y)$ derived from the sample space:

    a) Compute the entropy $H(X)$, representing the bank's initial uncertainty regarding fraud.

    b) Compute the conditional entropy $H(X \mid Y)$, representing the uncertainty remaining after the customer's risk class is revealed.

### Q1.3

By definition,  $E[X\mid Y]$ must be $\sigma(Y)$-measurable. This means it can depend only on the value of $Y$, not on the full outcome $\omega$. That is $E[X\mid Y]=g(Y)$ for some measurable map $g$.

###Q1.8

We calculate the Minimum Mean Square Error (MMSE), which gives the smallest possible average squared error when using the estimator $\mathbb{E}[X\mid Y]$.

The minimum MSE is

$$
\text{MSE}_{\min}
=
\mathbb{E}\left[(X-\mathbb{E}[X\mid Y])^2\right]
=
\mathbb{E}[\mathrm{Var}(X\mid Y)].
$$

Since $X$ is a Bernoulli random variable conditionally on $Y=y$, we use

$$
\mathrm{Var}(X\mid Y=y)
=
p(1-p).
$$

For $Y=0$, we have $p=\frac{1}{6}$, so

$$
\mathrm{Var}(X\mid Y=0)
=
\frac{1}{6}\left(1-\frac{1}{6}\right)
=
\frac{1}{6}\cdot\frac{5}{6}
=
\frac{5}{36}.
$$

For $Y=1$, we have $p=\frac{1}{2}$, so

$$
\mathrm{Var}(X\mid Y=1)
=
\frac{1}{2}\left(1-\frac{1}{2}\right)
=
\frac{1}{2}\cdot\frac{1}{2}
=
\frac{1}{4}.
$$

Now take the weighted average using the probabilities of $Y$:

$$
\mathbb{E}[\mathrm{Var}(X\mid Y)]
=
\mathrm{Var}(X\mid Y=0)P(Y=0)
+
\mathrm{Var}(X\mid Y=1)P(Y=1).
$$

Substituting the values,

$$
\mathbb{E}[\mathrm{Var}(X\mid Y)]
=
\frac{5}{36}(0.60)
+
\frac{1}{4}(0.40).
$$

Since $0.60=\frac{3}{5}$ and $0.40=\frac{2}{5}$,

$$
\mathbb{E}[\mathrm{Var}(X\mid Y)]
=
\frac{5}{36}\cdot\frac{3}{5}
+
\frac{1}{4}\cdot\frac{2}{5}.
$$

Therefore,

$$
\mathbb{E}[\mathrm{Var}(X\mid Y)]
=
\frac{1}{12}
+
\frac{1}{10}.
$$

Using a common denominator,

$$
\frac{1}{12}
+
\frac{1}{10}
=
\frac{5}{60}
+
\frac{6}{60}
=
\frac{11}{60}.
$$

Hence,

$$
\text{MSE}_{\min}
=
\frac{11}{60}
\approx 0.1833.
$$

So the lowest possible average squared error is

$$
\boxed{\frac{11}{60}\approx 0.1833}.
$$

### Q1.10 Quantifying Information Gain

First, we summarize the joint distribution of $X$ and $Y$.

From the sample space,

$$
P(X=0,Y=0)=P(\{\omega_1\})=0.50,
$$

$$
P(X=1,Y=0)=P(\{\omega_2\})=0.10,
$$

$$
P(X=0,Y=1)=P(\{\omega_3\})=0.20,
$$

and

$$
P(X=1,Y=1)=P(\{\omega_4\})=0.20.
$$

Therefore, the marginal probabilities are

$$
P(Y=0)=0.50+0.10=0.60,
$$

$$
P(Y=1)=0.20+0.20=0.40.
$$

Also,

$$
P(X=0)=0.50+0.20=0.70,
$$

and

$$
P(X=1)=0.10+0.20=0.30.
$$

Now we compute the conditional entropy $H(X\mid Y)$.

For $Y=0$,

$$
P(X=1\mid Y=0)=\frac{0.10}{0.60}=\frac{1}{6},
$$

and

$$
P(X=0\mid Y=0)=\frac{0.50}{0.60}=\frac{5}{6}.
$$

Thus,

$$
H(X\mid Y=0)
=
-\left[
\frac{1}{6}\log_2\left(\frac{1}{6}\right)
+
\frac{5}{6}\log_2\left(\frac{5}{6}\right)
\right].
$$

So,

$$
H(X\mid Y=0)\approx 0.650 \text{ bits}.
$$

For $Y=1$,

$$
P(X=1\mid Y=1)=\frac{0.20}{0.40}=\frac{1}{2},
$$

and

$$
P(X=0\mid Y=1)=\frac{0.20}{0.40}=\frac{1}{2}.
$$

Therefore,

$$
H(X\mid Y=1)
=
-\left[
\frac{1}{2}\log_2\left(\frac{1}{2}\right)
+
\frac{1}{2}\log_2\left(\frac{1}{2}\right)
\right].
$$

Hence,

$$
H(X\mid Y=1)=1 \text{ bit}.
$$

The conditional entropy is the weighted average:

$$
H(X\mid Y)
=
P(Y=0)H(X\mid Y=0)
+
P(Y=1)H(X\mid Y=1).
$$

Substitute the values:

$$
H(X\mid Y)
=
0.60(0.650)+0.40(1).
$$

So,

$$
H(X\mid Y)
=
0.390+0.400
=
0.790 \text{ bits}.
$$

Next, calculate the entropy of $X$:

$$
H(X)
=
-\left[
0.70\log_2(0.70)
+
0.30\log_2(0.30)
\right].
$$

Thus,

$$
H(X)\approx 0.881 \text{ bits}.
$$

Mutual information is given by

$$
I(X;Y)=H(X)-H(X\mid Y).
$$

Therefore,

$$
I(X;Y)
=
0.881-0.790
=
0.091 \text{ bits}.
$$

So the final values are

$$
H(X)=0.881 \text{ bits},
$$

$$
H(X\mid Y)=0.790 \text{ bits},
$$

and

$$
I(X;Y)=0.091 \text{ bits}.
$$

The percentage reduction in uncertainty is

$$
\frac{0.091}{0.881}\times 100
\approx 10.3\%.
$$

Therefore, knowing the risk class $Y$ reduces the uncertainty about fraud $X$ by about $10.3\%$. This shows that $Y$ gives useful information for predicting fraud.

# Q2) Conditional Density of the Multivariate Gaussian

We want to show that the conditional distribution of $\mathbf{X}$ given $\mathbf{Y}=\mathbf{y}$ is multivariate normal.

Let

$$
\mathbf{Z}
=
\begin{bmatrix}
\mathbf{X}\\
\mathbf{Y}
\end{bmatrix}
\sim
N\left(
\begin{bmatrix}
\boldsymbol{\mu}_X\\
\boldsymbol{\mu}_Y
\end{bmatrix},
\begin{bmatrix}
\boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY}\\
\boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY}
\end{bmatrix}
\right).
$$

The conditional density is defined as

$$
f_{\mathbf{X}\mid \mathbf{Y}}(\mathbf{x}\mid \mathbf{y})
=
\frac{f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}.
$$

Since $\mathbf{Y}$ is also multivariate normal,

$$
\mathbf{Y}\sim N(\boldsymbol{\mu}_Y,\boldsymbol{\Sigma}_{YY}).
$$

Hence,

$$
f_{\mathbf{Y}}(\mathbf{y})
=
\frac{1}{\sqrt{(2\pi)^m|\boldsymbol{\Sigma}_{YY}|}}
\exp\left[
-\frac{1}{2}
(\mathbf{y}-\boldsymbol{\mu}_Y)^T
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y)
\right].
$$

To simplify the joint density, define

$$
\boldsymbol{\Sigma}_{X\mid Y}
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

This matrix is called the Schur complement of $\boldsymbol{\Sigma}_{YY}$ in $\boldsymbol{\Sigma}$.

Using the block inverse formula, the quadratic term in the joint density can be separated as

$$
(\mathbf{z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{z}-\boldsymbol{\mu})
=
(\mathbf{x}-\boldsymbol{\mu}^*)^T
\boldsymbol{\Sigma}_{X\mid Y}^{-1}
(\mathbf{x}-\boldsymbol{\mu}^*)
+
(\mathbf{y}-\boldsymbol{\mu}_Y)^T
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y),
$$

where

$$
\boldsymbol{\mu}^*
=
\boldsymbol{\mu}_X
+
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y).
$$

Also, by the determinant formula for block matrices,

$$
|\boldsymbol{\Sigma}|
=
|\boldsymbol{\Sigma}_{YY}|
\,
|\boldsymbol{\Sigma}_{X\mid Y}|.
$$

Therefore, the joint density can be written as

$$
f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})
=
f_{\mathbf{X}\mid \mathbf{Y}}(\mathbf{x}\mid \mathbf{y})
f_{\mathbf{Y}}(\mathbf{y}).
$$

Dividing the joint density by $f_{\mathbf{Y}}(\mathbf{y})$, the terms involving only $\mathbf{y}$ cancel out. Thus,

$$
f_{\mathbf{X}\mid \mathbf{Y}}(\mathbf{x}\mid \mathbf{y})
=
\frac{1}{\sqrt{(2\pi)^n|\boldsymbol{\Sigma}_{X\mid Y}|}}
\exp\left[
-\frac{1}{2}
(\mathbf{x}-\boldsymbol{\mu}^*)^T
\boldsymbol{\Sigma}_{X\mid Y}^{-1}
(\mathbf{x}-\boldsymbol{\mu}^*)
\right].
$$

Hence,

$$
\mathbf{X}\mid \mathbf{Y}=\mathbf{y}
\sim
N\left(
\boldsymbol{\mu}_X
+
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y),
\;
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\right).
$$

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.

\mathbf{y}-\boldsymbol{\mu}_Y

\boldsymbol{\Sigma}^{-1}

\mathbf{x}-\boldsymbol{\mu}_X\\
\mathbf{y}-\boldsymbol{\mu}_Y
\end{bmatrix}.

$$
\mathbf{S}_{YY}
=
\boldsymbol{\Sigma}_{YY}

\boldsymbol{\Sigma}_{YX}
\boldsymbol{\Sigma}_{XX}^{-1}
\boldsymbol{\Sigma}_{XY}.

(\mathbf{x}-\boldsymbol{\mu}_X).

$$
|\boldsymbol{\Sigma}|

|\boldsymbol{\Sigma}_{XX}|
|\mathbf{S}_{YY}|.
$$

f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})

f_{\mathbf{X}}(\mathbf{x})

# 3. Conditional Density

To derive the conditional density of $\mathbf{X}$ given $\mathbf{Y}=\mathbf{y}$, we start from the definition

$$
f_{\mathbf{X}\mid \mathbf{Y}}(\mathbf{x}\mid \mathbf{y})
=
\frac{
f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})
}{
f_{\mathbf{Y}}(\mathbf{y})
}.
$$

From the previous decomposition using the Schur complement of $\boldsymbol{\Sigma}_{YY}$, the quadratic form in the joint density becomes

$$
Q_{\text{joint}}
=
(\mathbf{y}-\boldsymbol{\mu}_Y)^T
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y)
+
(\mathbf{x}-\boldsymbol{\mu}^*)^T
\mathbf{S}_{XX}^{-1}
(\mathbf{x}-\boldsymbol{\mu}^*),
$$

where

$$
\mathbf{S}_{XX}
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX},
$$

and

$$
\boldsymbol{\mu}^*
=
\boldsymbol{\mu}_X
+
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y).
$$

The marginal density of $\mathbf{Y}$ contains the quadratic term

$$
Q_{\text{marginal}}
=
(\mathbf{y}-\boldsymbol{\mu}_Y)^T
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y).
$$

When dividing the joint density by the marginal density, the terms involving only $\mathbf{y}$ cancel. Therefore,

$$
Q_{\text{conditional}}
=
Q_{\text{joint}}-Q_{\text{marginal}}
=
(\mathbf{x}-\boldsymbol{\mu}^*)^T
\mathbf{S}_{XX}^{-1}
(\mathbf{x}-\boldsymbol{\mu}^*).
$$

Now simplify the constants.

The joint density constant is

$$
\frac{1}{\sqrt{(2\pi)^{n+m}|\boldsymbol{\Sigma}|}},
$$

while the marginal density constant is

$$
\frac{1}{\sqrt{(2\pi)^m|\boldsymbol{\Sigma}_{YY}|}}.
$$

Their ratio becomes

$$
\frac{
1
}{
\sqrt{(2\pi)^{n+m}|\boldsymbol{\Sigma}|}
}
\cdot
\sqrt{
(2\pi)^m|\boldsymbol{\Sigma}_{YY}|
}.
$$

This simplifies to

$$
\frac{
1
}{
\sqrt{
(2\pi)^n
\frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}
}
}.
$$

Using the determinant identity

$$
|\boldsymbol{\Sigma}|
=
|\boldsymbol{\Sigma}_{YY}|
|\mathbf{S}_{XX}|,
$$

we obtain

$$
\frac{
1
}{
\sqrt{
(2\pi)^n
|\mathbf{S}_{XX}|
}
}.
$$

Combining the simplified constant with the remaining exponential term gives

$$
f_{\mathbf{X}\mid \mathbf{Y}}(\mathbf{x}\mid \mathbf{y})
=
\frac{
1
}{
\sqrt{
(2\pi)^n
|\mathbf{S}_{XX}|
}
}
\exp\left(
-\frac{1}{2}
(\mathbf{x}-\boldsymbol{\mu}^*)^T
\mathbf{S}_{XX}^{-1}
(\mathbf{x}-\boldsymbol{\mu}^*)
\right).
$$

Hence,

$$
(\mathbf{X}\mid \mathbf{Y}=\mathbf{y})
\sim
\mathcal{N}
\left(
\boldsymbol{\mu}_X
+
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y),
\;
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\right).
$$

Therefore, the conditional mean is

$$
\boldsymbol{\mu}_{X\mid Y}
=
\boldsymbol{\mu}_X
+
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y),
$$

and the conditional covariance is

$$
\boldsymbol{\Sigma}_{X\mid Y}
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

An important observation is that the conditional covariance does not depend on the observed value $\mathbf{y}$. This property is fundamental in Gaussian linear regression and Kalman filtering, where uncertainty reduction depends only on the covariance structure and not on the specific observation value.

# 4. Conditional Expectation

Let $g(y) = \mathbb{E}[X \mid Y=y]$. By the Disintegration Theorem, this is computed as the integral against the conditional PDF:
$$g(y) = \int_{\mathbb{R}^n} \mathbf{x} f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) \, d\mathbf{x}.$$
This integral is simply the first moment (mean) of a Gaussian random variable
$$(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) \sim \mathscr{N}(\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y), \quad \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}).$$


Thus:
$$g(y) = \mathbf{\mu}(\mathbf{y}) = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y).$$

Therefore:
$$\mathbb{E}[X \mid Y=y] = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y).$$

and hence
$$\mathbb{E}[X \mid Y] = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{Y} - \boldsymbol{\mu}_y)$$



# 5. The Conditional Variances

Since

$$
(\mathbf{X}\mid \mathbf{Y}=\mathbf{y})
\sim
N\left(
\boldsymbol{\mu}_X
+
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
(\mathbf{y}-\boldsymbol{\mu}_Y),
\;
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\right),
$$

we have

$$
\mathrm{Var}(\mathbf{X}\mid \mathbf{Y}=\mathbf{y})
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

This expression does not depend on the particular value of $\mathbf{y}$. Hence,

$$
\mathrm{Var}(\mathbf{X}\mid \mathbf{Y})
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

Therefore,

$$
\mathbb{E}\left[\mathrm{Var}(\mathbf{X}\mid \mathbf{Y})\right]
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

Thus the minimum mean square estimation error is

$$
\mathbb{E}
\left[
(\mathbf{X}-\mathbb{E}[\mathbf{X}\mid \mathbf{Y}])^2
\right]
=
\mathbb{E}\left[
\mathrm{Var}(\mathbf{X}\mid \mathbf{Y})
\right].
$$

So,

$$
\mathbb{E}
\left[
(\mathbf{X}-\mathbb{E}[\mathbf{X}\mid \mathbf{Y}])^2
\right]
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

Now apply the law of total variance:

$$
\mathrm{Var}(\mathbf{X})
=
\mathbb{E}[\mathrm{Var}(\mathbf{X}\mid \mathbf{Y})]
+
\mathrm{Var}(\mathbb{E}[\mathbf{X}\mid \mathbf{Y}]).
$$

Since

$$
\mathrm{Var}(\mathbf{X})
=
\boldsymbol{\Sigma}_{XX},
$$

and

$$
\mathbb{E}[\mathrm{Var}(\mathbf{X}\mid \mathbf{Y})]
=
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX},
$$

we get

$$
\mathrm{Var}(\mathbb{E}[\mathbf{X}\mid \mathbf{Y}])
=
\boldsymbol{\Sigma}_{XX}
-
\left(
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\right).
$$

Therefore,

$$
\mathrm{Var}(\mathbb{E}[\mathbf{X}\mid \mathbf{Y}])
=
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

So the variance decomposition is

$$
\boldsymbol{\Sigma}_{XX}
=
\left(
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}
\right)
+
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

The unexplained variance, or estimation error, is

$$
\boldsymbol{\Sigma}_{XX}
-
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

The explained variance is

$$
\boldsymbol{\Sigma}_{XY}
\boldsymbol{\Sigma}_{YY}^{-1}
\boldsymbol{\Sigma}_{YX}.
$$

Hence, the covariance of $\mathbf{X}$ is split into the part explained by $\mathbf{Y}$ and the remaining unexplained part after observing $\mathbf{Y}$.

## 6.1 The Differential Entropy of a Multivaraite Gaussian

Let

$$
\mathbf{Z}
=
\begin{bmatrix}
\mathbf{X}\\
\mathbf{Y}
\end{bmatrix}
\in \mathbb{R}^{n+m}
$$

be a multivariate normal random vector with mean $\boldsymbol{\mu}$ and covariance matrix $\boldsymbol{\Sigma}$.

The joint density is

$$
f_{\mathbf{X},\mathbf{Y}}(\mathbf{z})
=
\frac{1}{\sqrt{(2\pi)^{n+m}|\boldsymbol{\Sigma}|}}
\exp\left(
-\frac{1}{2}
(\mathbf{z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{z}-\boldsymbol{\mu})
\right).
$$

The differential entropy is defined as

$$
h(\mathbf{Z})
=
-\mathbb{E}\left[\ln f_{\mathbf{X},\mathbf{Y}}(\mathbf{Z})\right].
$$

First, take the natural logarithm of the density:

$$
\ln f_{\mathbf{X},\mathbf{Y}}(\mathbf{z})
=
-\frac{1}{2}
\ln\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right)
-
\frac{1}{2}
(\mathbf{z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{z}-\boldsymbol{\mu}).
$$

Therefore,

$$
-\ln f_{\mathbf{X},\mathbf{Y}}(\mathbf{z})
=
\frac{1}{2}
\ln\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right)
+
\frac{1}{2}
(\mathbf{z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{z}-\boldsymbol{\mu}).
$$

Taking expectation gives

$$
h(\mathbf{Z})
=
\frac{1}{2}
\ln\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right)
+
\frac{1}{2}
\mathbb{E}\left[
(\mathbf{Z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{Z}-\boldsymbol{\mu})
\right].
$$

Now evaluate the expectation of the quadratic form. Using the trace identity,

$$
(\mathbf{Z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{Z}-\boldsymbol{\mu})
=
\mathrm{Tr}\left[
\boldsymbol{\Sigma}^{-1}
(\mathbf{Z}-\boldsymbol{\mu})
(\mathbf{Z}-\boldsymbol{\mu})^T
\right].
$$

Hence,

$$
\mathbb{E}\left[
(\mathbf{Z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{Z}-\boldsymbol{\mu})
\right]
=
\mathrm{Tr}\left[
\boldsymbol{\Sigma}^{-1}
\mathbb{E}\left[
(\mathbf{Z}-\boldsymbol{\mu})
(\mathbf{Z}-\boldsymbol{\mu})^T
\right]
\right].
$$

Since

$$
\mathbb{E}\left[
(\mathbf{Z}-\boldsymbol{\mu})
(\mathbf{Z}-\boldsymbol{\mu})^T
\right]
=
\boldsymbol{\Sigma},
$$

we get

$$
\mathbb{E}\left[
(\mathbf{Z}-\boldsymbol{\mu})^T
\boldsymbol{\Sigma}^{-1}
(\mathbf{Z}-\boldsymbol{\mu})
\right]
=
\mathrm{Tr}(\boldsymbol{\Sigma}^{-1}\boldsymbol{\Sigma}).
$$

Thus,

$$
\mathrm{Tr}(\boldsymbol{\Sigma}^{-1}\boldsymbol{\Sigma})
=
\mathrm{Tr}(\mathbf{I}_{n+m})
=
n+m.
$$

Substitute this into the entropy expression:

$$
h(\mathbf{Z})
=
\frac{1}{2}
\ln\left((2\pi)^{n+m}|\boldsymbol{\Sigma}|\right)
+
\frac{n+m}{2}.
$$

Since

$$
\frac{n+m}{2}
=
\frac{1}{2}\ln(e^{n+m}),
$$

we can combine the logarithms:

$$
h(\mathbf{Z})
=
\frac{1}{2}
\ln\left((2\pi)^{n+m}e^{n+m}|\boldsymbol{\Sigma}|\right).
$$

Therefore,

$$
h(\mathbf{Z})
=
\frac{1}{2}
\ln\left((2\pi e)^{n+m}|\boldsymbol{\Sigma}|\right).
$$

## 6.2 Conditional Entropy

The conditional entropy $h(\mathbf{X}\mid\mathbf{Y})$ represents the average amount of uncertainty remaining in $\mathbf{X}$ after $\mathbf{Y}$ is known. For a multivariate Gaussian, this is determined by the varaince of the contional distribution
$$(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) \sim \mathscr{N}(\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y), \quad \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX})$$

The conditional covariance $\Sigma_{X|Y}$ covariance is thus given by $$\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}} = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}.$$  
Since $\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}}$ is independent of the specific value $y$, the uncertainty is constant across all observations of $Y$.  The differential entropy for this conditional distribution is thus :
$$h(\mathbf{X}\mid\mathbf{Y}) = \frac{1}{2} \ln((2\pi e)^n |\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}).$$

## 6.3 The Entropy of the Conditional Expectation

From the law of total variance we have seen that the explained varaince is:
$\text{Var}(\mathbb{E}[\mathbf{X}\mid\mathbf{Y}]) = \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$. Thus the differential entropy of $\mathbb{E}(\mathbf{X}\mid\mathbf{Y})$is:
$$h(\mathbb{E}(\mathbf{X}\mid\mathbf{Y})) = \frac{1}{2} \ln((2\pi e)^n |\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}|).$$

## 6.4 Mutual Information

Mutual information is given by:

$$I(\mathbf{X};\mathbf{Y}) = h(\mathbf{X}) - h(\mathbf{X} \mid \mathbf{Y}).$$

Substitute the entropy formulas for the marginal and conditional distributions derived in the document:

$$I(\mathbf{X};\mathbf{Y}) = \frac{1}{2} \ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{XX}|\right) - \frac{1}{2} \ln\left((2\pi e)^n |\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}}|\right) \text{.}$$

Using the property of logarithms ($\ln A - \ln B = \ln \frac{A}{B}$), the constants $(2\pi e)^n$ cancel out:

$$I(\mathbf{X};\mathbf{Y}) = \frac{1}{2} \ln \left( \frac{|\boldsymbol{\Sigma}_{XX}|}{|\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}}|} \right) \text{.}$$

Substituting the expression for the conditional covariance $\boldsymbol{\Sigma}_{\mathbf{X}|\mathbf{Y}} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$:

$$I(\mathbf{X};\mathbf{Y}) = \frac{1}{2} \ln \left( \frac{|\boldsymbol{\Sigma}_{XX}|}{|\boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}|} \right).$$

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**

    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)=\Sigma_{XY}$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

**5. Information Theory: Differential Entropy**
* Compute the differential entropy $h(X)$.
* Compute the conditional differential entropy $h(X \mid Y)$.
* *Note: The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.*

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?
* How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).



## Answer

Since $X$ and $Q$ are independent normal random variables, and

$$
Y=X+Q,
$$

the random vector

$$
\begin{pmatrix}
X\\
Y
\end{pmatrix}
$$

is jointly normal.

First, calculate the mean of $Y$:

$$
E[Y]
=
E[X+Q]
=
E[X]+E[Q]
=
0+0
=
0.
$$

Since $X$ and $Q$ are independent,

$$
\mathrm{Var}(Y)
=
\mathrm{Var}(X+Q)
=
\mathrm{Var}(X)+\mathrm{Var}(Q).
$$

Thus,

$$
\mathrm{Var}(Y)
=
1+4
=
5.
$$

Now compute the covariance between $X$ and $Y$:

$$
\mathrm{Cov}(X,Y)
=
\mathrm{Cov}(X,X+Q).
$$

Using linearity of covariance,

$$
\mathrm{Cov}(X,Y)
=
\mathrm{Cov}(X,X)+\mathrm{Cov}(X,Q).
$$

Since $X$ and $Q$ are independent,

$$
\mathrm{Cov}(X,Q)=0.
$$

Also,

$$
\mathrm{Cov}(X,X)=\mathrm{Var}(X)=1.
$$

Therefore,

$$
\mathrm{Cov}(X,Y)=1.
$$

So the joint distribution is

$$
\begin{pmatrix}
X\\
Y
\end{pmatrix}
\sim
\mathcal{N}
\left(
\begin{pmatrix}
0\\
0
\end{pmatrix},
\begin{pmatrix}
1 & 1\\
1 & 5
\end{pmatrix}
\right).
$$

The random variable $X$ is not $\sigma(Y)$-measurable because $X$ cannot be written as a deterministic function of $Y$. For one observed value $Y=y$, there can be many possible values of $X$ and $Q$ satisfying

$$
y=x+q.
$$

So observing $Y$ does not reveal $X$ exactly.

For jointly normal random variables, the conditional expectation is

$$
E[X\mid Y=y]
=
\mu_X
+
\frac{\mathrm{Cov}(X,Y)}{\mathrm{Var}(Y)}
(y-\mu_Y).
$$

Substituting the values,

$$
E[X\mid Y=y]
=
0+\frac{1}{5}(y-0).
$$

Hence,

$$
E[X\mid Y=y]
=
\frac{1}{5}y.
$$

Therefore,

$$
E[X\mid Y]
=
\frac{1}{5}Y
=
0.2Y.
$$

This means that only $20\%$ of the observed value of $Y$ is used to estimate $X$, because much of the variation in $Y$ comes from the noise variable $Q$.

Now define the residual error

$$
Z=X-0.2Y.
$$

Its mean is

$$
E[Z]
=
E[X]-0.2E[Y]
=
0-0
=
0.
$$

Next, check that the residual is uncorrelated with $Y$:

$$
\mathrm{Cov}(Z,Y)
=
\mathrm{Cov}(X-0.2Y,Y).
$$

Thus,

$$
\mathrm{Cov}(Z,Y)
=
\mathrm{Cov}(X,Y)-0.2\mathrm{Var}(Y).
$$

Using $\mathrm{Cov}(X,Y)=1$ and $\mathrm{Var}(Y)=5$,

$$
\mathrm{Cov}(Z,Y)
=
1-0.2(5)
=
1-1
=
0.
$$

So the estimation error is orthogonal to the observation $Y$.

Now compute the differential entropy.

For a normal random variable with variance $\sigma^2$,

$$
h(N(\mu,\sigma^2))
=
\frac{1}{2}\ln(2\pi e\sigma^2).
$$

Since $\mathrm{Var}(X)=1$,

$$
h(X)
=
\frac{1}{2}\ln(2\pi e).
$$

Therefore,

$$
h(X)
\approx 1.4189 \text{ nats}.
$$

The conditional variance is

$$
\mathrm{Var}(X\mid Y)
=
\mathrm{Var}(X)
-
\frac{\mathrm{Cov}(X,Y)^2}{\mathrm{Var}(Y)}.
$$

Substituting the values,

$$
\mathrm{Var}(X\mid Y)
=
1-\frac{1^2}{5}
=
1-\frac{1}{5}
=
\frac{4}{5}.
$$

So,

$$
\mathrm{Var}(X\mid Y)=0.8.
$$

Hence,

$$
h(X\mid Y)
=
\frac{1}{2}\ln(2\pi e\cdot 0.8).
$$

Therefore,

$$
h(X\mid Y)
\approx 1.3073 \text{ nats}.
$$

The mutual information is

$$
I(X;Y)
=
h(X)-h(X\mid Y).
$$

Thus,

$$
I(X;Y)
=
\frac{1}{2}\ln(2\pi e)
-
\frac{1}{2}\ln(2\pi e\cdot 0.8).
$$

Simplifying,

$$
I(X;Y)
=
\frac{1}{2}\ln\left(\frac{1}{0.8}\right).
$$

Therefore,

$$
I(X;Y)
=
\frac{1}{2}\ln(1.25)
\approx 0.1116 \text{ nats}.
$$

The variance is reduced from

$$
1
$$

to

$$
0.8.
$$

So the percentage reduction in variance is

$$
\frac{1-0.8}{1}\times 100
=
20\%.
$$

If the manager uses the naive estimator

$$
\widehat{X}=Y,
$$

then the error is

$$
X-Y=X-(X+Q)=-Q.
$$

Thus,

$$
E[(X-Y)^2]
=
E[Q^2].
$$

Since $\mathrm{Var}(Q)=4$ and $E[Q]=0$,

$$
E[Q^2]=4.
$$

So the naive estimator has error variance

$$
4.
$$

However, using the conditional expectation

$$
\widehat{X}=E[X\mid Y]=0.2Y,
$$

the error variance is only

$$
0.8.
$$

Therefore, $0.2Y$ is much better than using $Y$ directly. This is a shrinkage estimator because it pulls the noisy observation $Y$ toward the mean value $0$.

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import multivariate_normal, norm

# --- 1. Parameters for the Spindle Problem ---
# X: Temperature ~ N(0, 1)
# Q: Torque ~ N(0, 4)
# Y: Power = X + Q => Y ~ N(0, 5)
mu_x, mu_y = 0, 0
var_x, var_q = 1, 4
var_y = var_x + var_q
cov_xy = var_x  # Cov(X, X+Q) = Var(X) = 1
rho = cov_xy / (np.sqrt(var_x) * np.sqrt(var_y))

# Shrinkage Factor (Gain)
k = cov_xy / var_y  # 1/5 = 0.2

# --- 2. Generate Data for Visualization ---
# Sample points to show the "Cloud" of data
np.random.seed(42)
n_samples = 1000
x_samples = np.random.normal(mu_x, np.sqrt(var_x), n_samples)
q_samples = np.random.normal(0, np.sqrt(var_q), n_samples)
y_samples = x_samples + q_samples
x_hat = k * y_samples  # E[X|Y]
residual = x_samples - x_hat

# Grids for Contours
x_grid = np.linspace(-4, 4, 200)
y_grid = np.linspace(-8, 8, 200)
X_m, Y_m = np.meshgrid(x_grid, y_grid)
pos = np.dstack((X_m, Y_m))
Z = multivariate_normal([mu_x, mu_y], [[var_x, cov_xy], [cov_xy, var_y]]).pdf(pos)

# --- 3. Create Simulation Plot ---
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "Joint Distribution: X vs Y<br><sup>The Line of Best Estimate (Shrinkage)</sup>",
        "Estimation Error (Residual) vs Y<br><sup>Visualizing the Orthogonality Principle</sup>"
    )
)

# --- LEFT PLOT: Joint X, Y and the Estimator ---
# Density Contour
fig.add_trace(go.Contour(
    x=x_grid, y=y_grid, z=Z, colorscale='Blues', showscale=False, opacity=0.6,
    hoverinfo='skip'
), row=1, col=1)

# Scatter of actual realizations
fig.add_trace(go.Scatter(
    x=x_samples, y=y_samples, mode='markers',
    marker=dict(color='rgba(0,0,0,0.1)', size=3),
    name="Realized (x, y)"
), row=1, col=1)

# The Best Estimate Line E[X|Y] = 0.2Y
y_line = np.linspace(-8, 8, 10)
fig.add_trace(go.Scatter(
    x=k * y_line, y=y_line,
    line=dict(color='Red', width=4),
    name="E[X|Y] = 0.2Y (Optimal)"
), row=1, col=1)

# The Naive Estimator Line X = Y
fig.add_trace(go.Scatter(
    x=y_line, y=y_line,
    line=dict(color='Black', width=2, dash='dash'),
    name="X = Y (Naive)"
), row=1, col=1)

# --- RIGHT PLOT: Orthogonality (Residual vs Y) ---
fig.add_trace(go.Scatter(
    x=y_samples, y=residual, mode='markers',
    marker=dict(color='Purple', opacity=0.4, size=4),
    name="Residual (X - 0.2Y)"
), row=1, col=2)

# Horizontal line at 0 to show no correlation
fig.add_trace(go.Scatter(
    x=y_line, y=np.zeros_like(y_line),
    line=dict(color='Black', width=2),
    name="Zero Mean Error"
), row=1, col=2)

# --- Layout and Styling ---
fig.update_layout(
    height=600,
    title_text="Spindle Motor Diagnostic Simulation: Extracting Signal from Noise",
    template="plotly_white",
    hovermode='x'
)

fig.update_xaxes(title_text="True Temperature (X)", row=1, col=1)
fig.update_yaxes(title_text="Observed Power (Y)", row=1, col=1)
fig.update_xaxes(title_text="Observed Power (Y)", row=1, col=2)
fig.update_yaxes(title_text="Residual Error (Z)", row=1, col=2)

fig.show()
```

The simulation results provide a powerful bridge between abstract probability theory and practical industrial diagnostics. Here is a detailed breakdown of what these results physically and mathematically imply:

1. The Shrinkage Principle (The "Conservative" Estimate)
In the left plot, we compare the **Naive Estimator** ($X=Y$) with the **Optimal Estimator** ($\mathbb{E}[X|Y] = 0.2Y$).
* **The Problem:** Because the noise (Torque $Q$, $\sigma^2=4$) is four times stronger than the signal (Temperature $X$, $\sigma^2=1$), the raw sensor reading $Y$ is mostly "polluted."
* **The Result:** The optimal estimator "shrinks" the measurement by 80%. If the power meter spikes to $Y=5$, the math tells us that, statistically, only $1$ unit of that spike is likely due to temperature, while $4$ units are likely just material hardness.
* **Physical Meaning:** $\mathbb{E}[X|Y]$ prevents **False Alarms**. A naive system would trigger an emergency cooling shut-off every time the tool hits a hard knot in the wood; the "shrunk" estimator recognizes that $Q$ is the more likely culprit for large spikes.

2. The Orthogonality Principle (Efficiency)
The right plot shows the **Residual Error ($Z$)** against the **Observation ($Y$)**.
* **The Observation:** The cloud of points is perfectly horizontal (zero correlation).
* **The Meaning:** This proves that our estimate $0.2Y$ has "sucked out" every possible bit of information that $Y$ had to offer.
* **The Proof:** If the plot showed a diagonal trend, it would imply that some part of the error could still be predicted by $Y$. Since the correlation is zero, the error is purely "white noise" relative to our sensor. We have achieved the **Minimum Mean Square Error (MMSE)**.

3. Information Geometry & Sigma-Algebras
The contour plot visualizes why $X$ is not $\sigma(Y)$-measurable:
* Look at the vertical spread for any single $Y$ value. Even if you fix $Y=0$, $X$ still has a spread (a conditional variance of $0.8$).
* Because the sensor $Y$ maps a 2D reality $(X, Q)$ down to a 1D line, information is lost. The "Best Estimate" is the vertical center-point of that spread, but the "thickness" of the cloud represents the **irreducible uncertainty** ($h(X|Y)$) that the sensor cannot resolve.

4. Variance Reduction and Mutual Information
* **Initial State:** Before measuring, our uncertainty (variance) about the temperature was **1.0**.
* **After Measurement:** Our uncertainty dropped to **0.8**.
* **Information Gain:** We gained **$0.1116$ nats** ($I(X;Y)$).
* **The "Why":** This reduction ($20\%$) is the physical value of the sensor. If we wanted a higher $I(X;Y)$, we wouldn't need a "better" version of $Y$; we would need a **different** sensor (like an acoustic sensor) that is less correlated with $Q$.
